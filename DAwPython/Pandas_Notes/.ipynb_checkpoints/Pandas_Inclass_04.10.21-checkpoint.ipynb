# recap for Apply - Transform - Map

```python
np.random.seed(42)
df3 = pd.DataFrame(  np.random.randn(5,4), 
                   index = ['A', 'B', 'C', 'D', 'E'], 
                   columns = 'W X Y Z'.split()          )
df3
```

```python
df6 = pd.DataFrame({'groups': ['A', 'B', 'C', 'A', 'B', 'C','A', 'B', 'C'],
                   'var1': [10,23,33,22,11,99,40,20,15],
                   'var2': [100,253,333,262,111,969, 232,343,123]})
df6
```

```python
df6[["var1","var2"]].transform( lambda x: (x-df6["var1"].min())/(df6["var1"].max()-df6["var1"].max()) )  # make minmaxscaler
```

```python
df6["var1"].transform( lambda x: (x-df6["var1"].min())/(df6["var1"].max()-df6["var1"].max()) )
```

```python
df6[["var1","var2"]].transform( lambda x: (x-x.min())/(x.max()-x.min()) ) # make minmaxscaler
```

```python
df6[["var1","var2"]].apply( lambda x: (x-x.min())/(x.max()-x.min()) )
```

```python
df6.groupby('groups')[["var1","var2"]].transform(np.mean)
```

```python
df6.groupby('groups')[["var1","var2"]].apply(np.mean)
```

```python
df6.groupby('groups')[["var1","var2"]].mean()
```

```python
# transform sütun bazında değişkenler üzerinde vektörül bazda bir işlem yapıyor. Missing value larda transform kullanacağız. Satır satır işlem yaptığından dolayı mean, sum, median gibi func burada kullanamayız.
# apply ise dataframe in değişkenleri üzerinde gezinen ve aggregate ile kullanılan bir fonksiyondur. 
```

# pivot table

```python
titanic = sns.load_dataset('titanic')
titanic                                   
```

```python
titanic.describe().T
```

```python
titanic.describe(include="all")
```

```python
titanic.info()
```

```python
titanic.groupby(["sex","pclass"])[["survived"]].mean()
```

```python
titanic.pivot_table(values="survived",index="sex",columns="pclass")
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean").unstack()
```

```python
titanic.groupby(["who","sex","pclass"])[["survived"]].mean().unstack()
```

```python
titanic.groupby(["who","sex","pclass"])[["survived"]].mean().unstack().unstack()
```

```python
titanic.pivot_table(values="survived",index=["sex","who"],columns="pclass")
```

```python
titanic.pivot_table(values="survived",index=["sex","who"],columns="pclass",aggfunc="count")
```

```python
titanic.groupby(["who","sex","pclass"])[["survived"]].mean()
```

```python
titanic.groupby(["who","sex","pclass"])[["survived"]].mean().unstack(-1)  # stack içindeki sıra grup by içindeki liste göre 0,1,2 veya tersten -1,-2,-3 dür.
```

```python
df = pd.DataFrame({'A':[1, 2, np.nan],
                  'B':[5, np.nan, np.nan],
                  'C':[1, 2, 3]})
df
```

```python
df.isnull()  # NaN değerleri True olarak veriyor.
```

```python
df.isna()
```

```python
df.notnull()# NaN olmayan değerler için True, olanlar için False döndürür.
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df2 = df.append([np.nan]).reset_index(drop=True)
df2
```

```python
df2.dropna()  # axis = 0 default olarak atadığından va 0 column de hepsi NaN olduğundan hepsini sildib
```

```python
df2.dropna(how="all")  # hepsi NaN olanı düşürdü.
```

```python
df2
```

```python
df2.dropna(thresh=1,axis=0)  # Keep only the rows with at least 1 non-NA values.
```

```python
df2.dropna(thresh=2,axis=0) # Keep only the rows with at least 2 non-NA values.
```

```python
df2.dropna(thresh=3,axis=0)  # Keep only the rows with at least 3 non-NA values.
```

```python
df2.fillna(-1)   # NaN olanları ne ile dolduracağımızı seçiyoruz.
```
