# Take a look at correlation between variables and also see if there are some outliers in any of the columns.

```python
df.corr()  # yapılacak tüm analizlerde korelasyona bakacağız.
```

```python
df.describe()
```

```python
# describe bakınca max değerlerde çok uçuk değerler olması outlier olduğunun bir göstergesi olabilir.
```

```python
#age column
#scipy.stats.iqr
Q1 = df.age.quantile(0.25)
Q3 = df.age.quantile(0.75)
IQR = Q3-Q1
IQR
```

```python
lower_lim_age = Q1 - 1.5*IQR
lower_lim_age  # sonuca göre alt tarafta outlier yok.
```

```python
upper_lim_age = Q3 + 1.5*IQR
upper_lim_age
```

```python
df[df["age"] >=   upper_lim_age  ]["age"].value_counts()
```

```python
fig ,ax = plt.subplots(2,3,figsize=(20,7))
x = 0
y = 0
for column in df.select_dtypes('number').columns:
    
    df[column].plot(kind = "box", ax = ax[x,y])
    if y < 2:
        y += 1
    else:
        x += 1
        y = 0
# tüm numerik değişkenlerin boxplot grafiklerine bakarak outlier hakkınd bir fikir sahibi olabiliriz.
```

```python
df[(df["hours-per-week"] == 99)].sample(10)
```

```python
df["hours-per-week"].describe()
```

```python
#hours per week column
Q1 = df['hours-per-week'].quantile(0.25)
Q3 = df['hours-per-week'].quantile(0.75)
IQR_hour = Q3-Q1
IQR_hour
```

```python
lower_lim_hour = Q1 - 1.5*IQR_hour
lower_lim_hour
```

```python
upper_lim_hour = Q3 + 1.5*IQR_hour
upper_lim_hour
```

```python
# find outlier values which extends the upper limit in both age and hours-per-week columns
df[(df.age >= upper_lim_age) & (df["hours-per-week"] >= upper_lim_hour)]
```

 # What country has the highest percentage of people that earn >50K?
 ---


```python
country_salary_df = df.groupby('native-country')[['salary']].count() #salary i dataframe olarak aldık.
country_salary_df
```

```python
country_salary_df_riches = df[df.salary == '>50K'].groupby('native-country')[['salary']].count()
#önce filtreleme ile 50K da büyük olanları bulduk sonra ülkelere göre gruplandırdık. Buradan salary çektik. ve saydırdık. 
# country_salary_df_riches = df.groupby(["native-country"])[['salary']].apply(lambda x: (x == ">50K").sum())
country_salary_df_riches
```

```python
rich_per_country = (country_salary_df_riches/country_salary_df).sort_values('salary')
rich_per_country
```

```python
rich_per_country[rich_per_country.salary == rich_per_country.salary.max()]
```

```python
#use iloc here
rich_per_country.sort_values('salary',ascending=False).iloc[0]
```

```python
rich_per_country.sort_values('salary',ascending=False).iloc[[0]]
```
