```python
data = {'Company':['GOOGLE', 'GOOGLE', 'MICROSOFT', 'MICROSOFT', 'FACE', 'FACE'],
       'Person':['PB', 'Chandler', 'Ross', 'Joey', 'Rachel', 'Monica'],
       'Sales':[100, 140, 150, 99, 102, 110]}
```

```python
df4 = pd.DataFrame(data)
df4
```

```python
my_map = {"GOOGLE":"GOO","MICROSOFT":"MIC","FACE":"FB"}
```

```python
# map DataFrame uygulanmıyor. Sadece bir sütuna uygulayabiliyoruz.
# replace ise tüm DataFrame uygulanabiliyor.
```

```python
df4.Company = df4.Company.map(my_map)
df4
# df[‘Company’].map(lambda x : x[:3]) ilede yapılabilir.
```

```python
df4.Sales = df4.Sales.map(lambda x: x*100)
df4
```

```python
df4.Company.replace(to_replace=["GOO","MIC","FB"], value=["GOOGLE","MICROSOFT","FACE"], inplace=True)

df4
# df4.replace({"GOO":"GOOGLE", "MIC": "MICROSOFT", "F": "FACE" })
```

```python
df6 = pd.DataFrame({'groups': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'var1': [10,20,30,50,52,100],
                   'var2': [100,200,300,500,500,700]})
df6
```

```python
df6.groupby("groups")  #lasy object oluştu.
```

```python
df6.groupby("groups").sum()
```

```python
df6.groupby("groups")["var1"].sum()  # sum numeric değerlere işlem yapar.
```

```python
df6.groupby("groups").aggregate(["mean","max","min"])  # birden fazla kullanacaksak liste şeklinde vermeliyiz.
# agg veya aggregate yazılabilir.
```

```python
df6.groupby("groups").agg({"var1":("min","max"),"var2":"median"})  # farklı columnlere farklı işlemler atamak istediğimiz zaman kullanılır.
```

```python
df6[df6.var1>50]  # var1 in 50 den büyük olduğu yerleri getirir.
```

```python
df6.groupby("groups").sum()
```

```python
df6.groupby("groups").filter(lambda x: x["var1"].sum()>80)  # filter içine yazdığımız x["var1"] eğer agg func ile kullanmazsak hata verir.
# yukarıda gördüğümüz gibi 80 den büyük olan sadece C var. Buna ait tüm sonuçları getirdi.
```

```python
df6
```

```python
df6.groupby("groups").transform("mean")  # datafrome in formatını bozmadan A gördüğü her yere aynı değerleri döndürüyor. Yani o grubun mean yazıyor.
```

```python
df6.groupby("groups").transform("mean").drop_duplicates()
```

```python
df6.groupby("groups").transform(np.log) # her birvalue nin değer bazlı log aldı.
```

```python
df6.groupby("groups").apply(np.mean)
# apply transformdan farklı olarak formatı değiştirerek sankı duplicate yapılmış olarak tekrarsız getirir.
```

```python
df1 = pd.DataFrame([["a", 9, 25]] * 4, columns=["grp", 'P', 'Q'])
df2 = pd.DataFrame([["b", 9, 25]] * 3, columns=["grp", 'P', 'Q'])
df3 = pd.concat([df1, df2], ignore_index=True)
df3
```

```python
df3.groupby("grp").apply(sum) # apply da kontrol daha çok bizdedir. Daha çok seçenek vardır.
```

```python
df3.groupby("grp").transform(sum)  #transform missing value ları doldururken çok işe yarıyor. Apply ın daha özel halıdır.
```

```python
# DataFrame birleştirirken daha çok merge kullanılır. Çünkü daha çok seçeneği vardır.
# Apply ise işlem yaparken bize daha kolaylık sağlar.
# merge en yetenekli olan. Default olarak ilk DF in indekslerine bakarak birleştirme yapıyor. Default inner dır.
```

```python
left = pd.DataFrame({
   'id':[1,2,3,4,5],
   'Name': ['Arthur', 'Michael', 'Jason', 'David', 'Thomas'],
   'subject_id':['sub1','sub2','sub4','sub6','sub5']})
right = pd.DataFrame({
    'id':[1,2,3,4,6],
   'Name': ['Mehmet', 'Terry', 'Joseph', 'Matt', 'Bruce'],
   'subject_id':['sub2','sub4','sub3','sub6','sub5']})
```

```python
display(left,right)
```

```python
pd.merge(left,right,on="id") # id ler üzerinden birleştirdik. default olarak inner üzerinden işlem yaptı.
```

```python
pd.merge(left,right,on="id",how="outer")
```

```python
pd.merge(left,right,on="id",how="left")
```

```python
pd.merge(left,right,on="id",how="right")
```

```python
pd.merge(left,right,on=["id","subject_id"])
```

```python
# join de default olarak left alır. join çok fazla kullanılmıyor.
```

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index = ['K0', 'K1', 'K2'])
right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index = ['K0', 'K2', 'K3'])
```

```python
left
```

```python
right
```

```python
left.join(right)
```

```python
left.join(right)
```

```python
left.join(right, how="inner")
```
