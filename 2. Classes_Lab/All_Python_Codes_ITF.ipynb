Given a mapping of digits to letters (as in a phone number), and a digit string,return all possible letters the number could represent. You can assume each valid number in the mapping is a single digit.
For example if {“2”: [“a”, “b”, “c”], 3: [“d”, “e”, “f”], …} then “23” should return [“ad”, “ae”, “af”, “bd”, “be”, “bf”, “cd”, “ce”, “cf"].

```python
import random

def phone_letter(num):
    
    dictt = {2: 'abc', 3: 'def', 4: 'ghi', 5: 'jkl', 6: 'mno', 7: 'pqrs', 8: 'tuv', 9: 'wxyz'}
    
    times = 1
    for i in range(len(str(num))):
        times *= len(dictt[int(str(num)[i])])
    
    a = 0
    result_list = []
    while a < times:
        liste = []
        for i in range(len(str(num))):
            random_chr = random.choice(dictt[int(str(num)[i])])
            liste.append(random_chr)
        if liste not in result_list:
            result_list.insert(len(result_list), liste)
            a += 1
    result_list = [''.join(i) for i in result_list]
    result_list.sort()
    return result_list

phone_letter(642)
```

```python
def perfect_number(n):
    m = 10
    result = []
    while True:
        summ = sum([int(i) for i in str(m)])
        if summ == 10:
            result.append(m)
            if len(result) == n:
                break
            else:
                m += 1
        else:
            m += 1
    return result[-1]

perfect_number(2)
```

```python
def largest_product(array):    
    res = [[]]
    result = []
    for i in array:
        for j in res:
            res = res + [j + [i]]
    for i in res:
        if len(i) == 3:
            carpim = 1
            for j in i:
                carpim *= j
            result.append(carpim)
    return max(result)

largest_product([-10, -10, 5, 2])
```

```python
def pal(text):
    res, resa, son = [], [], []
    for i in range(len(text)):
        res.append(text[i:])
        resa.append(text[i::-1])

    for i in res:
        for j in resa:
            a = 0
            toplam = ''
            if i[a] == j[a]:
                while i[a] == j[a]:
                    toplam += i[a]
                    a += 1
                    if toplam == toplam[::-1]:
                        if toplam not in son:
                            son.append(toplam)
                    if a == len(i) or a == len(j):
                        break
    a = sorted(son, key=lambda x : len(x), reverse = True)
    return a[0]

pal('bananas')
```

```python
Find lcm (least common multiple) with divisors of three or more numbers
Examples:

lcm([12, 8, 16])  ===>  ([2, 2, 2, 2, 3], 48)
lcm([1, 2, 3, 4, 5, 6, 7, 8, 9])  ===>  ([2, 2, 2, 3, 3, 5, 7], 2520)
```

```python
from functools import reduce
def least_common(liste):
    a = 2
    res = []
    while True:
        b = False
        for i in range(len(liste)):
            if liste[i] % a == 0:
                b = True
                liste[i] = liste[i] // a
                liste.insert(i, liste[i])
                liste.pop(i + 1)
        if b:res.append(a)
        else:a += 1
        if liste.count(1) == len(liste):break
    result = reduce(lambda x, y: x * y, res)
    return (res, result)

least_common([12, 8, 16])
```

```python
def lcm(num_list):
    
    def multiplier(num):
        if num % i == 0: return int(num / i)
        else: return num

    def least_common_multiple(divisors):
        counter = 1
        for num in divisors: counter *= num; 
        return counter
                        
    divisors = []
    for i in range(2, max(num_list)+1):
        if max(num_list) == 1: break;
        while any([num % i == 0 for num in num_list]):
            num_list = list(map(multiplier, num_list))      
            divisors.append(i)

    return divisors, least_common_multiple(divisors)

print(lcm([12, 8, 16]))
```

```python
def next_letters(text):

    letters_list = list(text)
    new_text = ""

    while letters_list[-1:] == ["Z"]:
      new_text += "A"
      del letters_list[-1]

    if letters_list[-1:] == []:
      new_text += "A"
    else:
      new_text += chr(ord(letters_list[-1]) + 1)
      del letters_list[-1]
      for i in reversed(letters_list):
        new_text += i

    return "".join(reversed(new_text))

next_letters("CZADZZ")
```

```python
def chinese_puzzle(heads, legs):    
    for c in range(heads):
        for r in range(heads):
            if (2 * c) + (4 * r) == legs and c + r == heads:
                chic = c
                rab = r
    return f'Chickens: {chic} || Rabbits: {rab}'

chinese_puzzle(35, 94)
```

```python
def lucky_number(size, n):
    count = 1
    liste = list(range(1, size + 1))
    liste_2 = list(range(1, size + 1))
    while len(liste) >= liste_2[count]:
        deleted = []
        for i in liste:
            if (liste.index(i) + 1) % liste_2[count] == 0:
                deleted.append(liste.index(i))
        for i in deleted[::-1]:
            liste.pop(i)
        if liste_2[count + 1] in liste:
            count += 1
        else:
            count += 1
            while liste_2[count] not in liste:
                count += 1
    result = liste[n - 1]
    return result

lucky_number(120, 13)
```

A common way to simplify square roots is to repeatedly factor out perfect squares from the number underneath the square root. For example, if you need to simply sqrt(72), you can factor out perfect squares from 72 according to the following process:

sqrt(72)
72 is divisible by 4, so factor 4 out of 72:

sqrt(4⋅18)
The perfect square 4 can now be square rooted and pulled out of the square root:

2⋅sqrt(18)

Pull the 9 out, square root it, and simplify:

2⋅3⋅sqrt(2)

https://edabit.com/challenge/WyttgdGuQGaRBqhhP  Minimum Steps to a Palindrome

Given an incomplete palindrome as a string, return the minimum letters needed to be added on to the end to make the string a palindrome.

Examples
min_palindrome_steps("race") ➞ 3
# Add 3 letters: "car" to make "racecar"

min_palindrome_steps("mada") ➞ 1
# Add 1 letter: "m" to make "madam"

min_palindrome_steps("mirror") ➞ 3
# Add 3 letters: "rim" to make "mirrorrim"
Notes
Trivially, words which are already palindromes should return 0.

```python
def min_palindrome_steps(txt):
    if txt == txt[::-1]:
        return 0
    else:
        a = list(txt)
        a = a + [a[0]]
        count = 1
        if a == a[::-1]:
            return count
        else:
            for i in range(1,len(a)):
                count += 1
                a.insert(-i,a[i])
                if a == a[::-1]:
                    break
    return count
```

```python
text = input('Enter a text > ')
text2 = text
rev = text[::-1]
sayac = 0
for i in range(len(text2)):
    text = text[1:]
    rev = rev[:-1]
    sayac += 1
    if text == rev:
        break
which = text2[:sayac]
which_reverse = which[::-1]
print(f'''
We need minimum {sayac} letters to be added to the end of {text2},
These {sayac} letter is/are {which_reverse}.
Thus, our palindrom is {text2 + which_reverse}.''')
```

Examples
is_harshad(75) ➞ False
# 7 + 5 = 12
# 75 is not exactly divisible by 12
 
is_harshad(171) ➞ True
# 1 + 7 + 1 = 9
# 9 exactly divides 171
 
is_harshad(481) ➞ True

is_harshad(89) ➞ False

is_harshad(516) ➞ True

is_harshad(200) ➞ True
Notes
A recursive version of this challenge can be found in here.

```python
def is_harshad(n):
    a = []
    t = n
    while t > 0:
        a.append(t%10)
        t = t//10
    return True if n % sum(a) == 0 else False
```

```python
anagrams = {}
words = ["eat", "tea", "tan", "ate", "nat", "bat"]
for word in words:
    sorted_word = "".join(sorted(word))
    if sorted_word in anagrams:
        anagrams[sorted_word].append(word)
    else:
        anagrams[sorted_word] = [word]
print(list(anagrams.values()))
```

```python
def is_isogram(txt):
    txt = txt.lower()
    for letter in txt:
        if txt.count(letter) > 1:
            return False
    return True
```

Your task is to sort a given string. Each word in the string will contain a single number. This number is the position the word should have in the result.
Note: Numbers can be from 1 to 9. So 1 will be the first word (not 0).
If the input string is empty, return an empty string. The words in the input String will only contain valid consecutive numbers.
Examples
"is2 Thi1s T4est 3a"  -->  "Thi1s is2 3a T4est"
"4of Fo1r pe6ople g3ood th5e the2"  -->  "Fo1r the2 g3ood 4of th5e pe6ople"

```python
def order(list1):
    w = list1.split()
    newsen = ""
    i = 1
    while i < len(w) + 1:
        for word in w:
            if str(i) in word:
                newsen += word + ' '
        i += 1
    return newsen.strip()
```
