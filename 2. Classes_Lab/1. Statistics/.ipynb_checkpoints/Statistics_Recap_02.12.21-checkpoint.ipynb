## Exercise 1

A salesperson has found that the probability of a sale on a single contact is approximately 0.3. If the salesperson contacts 10 prospects, what is the approximate probability of making at least one sale?

Bu soruyu görünce aklımıza binominal dağılım geliyor. Bir başarı değeri var. stats.binom.pmf(k,n,p) bize hiç satış yapamama olasılığını verir. Bu bulduğumuz değeri 1 den çıkarırsak bize en az 1 den fazla satış yapma olasılığını verir. 

pmf : discrete olasılıkları hesaplamak için kullanılır.
cdf : continuous olasılıkları hesaplama için kullanılır.
ppf : olasılıktan x veya k değerine gidiyoruz.

```python
n = 10   # deneme sayısı
k = 0    # random variable P(X>=1)=1-P(X=0)
p = 0.3  # success oranı
```

```python
1- stats.binom.pmf(k,n,p)  # bulduğumuz 0 satış yapma olasılığını 1 den çıkarırız.
```

## Exercise 2

The cycle time for trucks hauling concrete to a highway construction site is uniformly distributed over the interval 50 to 70 minutes.

What is the probability that the cycle time exceeds 65 minutes if it is known that the cycle time exceeds 55 minutes?

Burada bir uniform dist söz konusu. Bilinen bir durum var. Bunun üzerinden bir sonuç istiyor. Yani 55 geçme olasılığı bilinirken 65 geçme olasılığını soruyor. Conditional Probability söz konusu.

```python
a = 50
b = 70

unifDist = stats.uniform(a, b-a)
```

```python
# Event A

pA = 1-unifDist.cdf(65)
pA
```

```python
# Event B

pB = 1-unifDist.cdf(55)
pB
```

```python
pA/pB  # pA|pB = (pA Kesişim pB)/pB
```

```python
## 2.Yöntem

# The probability that the cycle time exceeds 55 minutes is:
x = 1 - stats.uniform.cdf(55, loc=50, scale=20)

# The probability that the cycle time exceeds 65 minutes is:
y = 1 - stats.uniform.cdf(65, loc=50, scale=20)

# Given that the cycle time exceeds 55 minutes, the probabily that it exceeds 65 minutes is:
y / x
```

## Exercise 5

What is the normal body temperature for healthy humans? A random sample of 130 healthy human body temperatures provided by Allen Shoemaker yielded 98.25 degrees and standard deviation 0.73 degrees. Give a 99% confidence interval for the average body temperature of healthy people.

```python
xbar = 98.25  # sample mean
std = 0.73
n = 130  # sample size

conflevel = 0.995  # bunun için z value z table dan 0.005 için yani olasılıktan rv ye gideriz.
```

```python
stderror = std/np.sqrt(n)
stderror
```

```python
z = 0.995 # veya negatifi 0.005 tir
z = stats.norm.ppf(z)
z
```

```python
moe = z*stderror
moe
```

```python
lower = xbar - moe
lower
```

```python
upper = xbar + moe
upper
```

```python
# 2. yöntem:

stats.norm.interval(conflevel, loc=xbar, scale=stderror)  # interval hesabında std yerine std error veriyoruz.
```

```python
# 3. yöntem:

stats.t.interval(alpha=conflevel, df=n-1, loc=xbar, scale=std/np.sqrt(n))  # std/np.sqrt(n)=stderror

# t dist. sonuç yaklaşık olarak aynı çıktı fark t dist. başlangıcta tail kısmında daha basık olmasından ve 
# df arttıkça norm.dist. yaklaşmasından kaynaklanmaktadır.
```

⭐**Investigate the given task by using Independent Samples t Test.**

___🚀First, check the normality for both groups. *Use scipy.stats.shapiro*

!!! önemli:

Burada öncelikle assumption çheck dediğimiz grupların kendi içinde normal bir dağılım sergileme durumunu shapiro testi ile varyanslarının eşit olması durumunu ise levene testi ile inceleyeceğim. Daha sonra ise asıl hipoteze gelerek grupların ortalamalarının eşit olup olmadığına bakıyorum.

Öncelikle shapiro testi ile her iki grubun normal bir dağılım sergileyip sergilemediğine bakıyoruz. Ve sonuçta iki grup içinde H0 hipotezini reddetmekte başarısız olduğumdan dolayı grupların normal bir dağılım sergilediğini söyleyebiliriz.

```python
#Check the normality for male group

stat, p = stats.shapiro(df[df["gender"]=="Male"]["temperature"])

print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Sample looks Gaussian (fail to reject H0)')
else:
	print('Sample does not look Gaussian (reject H0)')
```

```python
#Check the normality for female group

stat, p = stats.shapiro(df[df["gender"]=="Female"]["temperature"])

print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('Sample looks Gaussian (fail to reject H0)')
else:
	print('Sample does not look Gaussian (reject H0)')
```

___🚀Test the assumption of homogeneity of variance
*Hint: Levene’s Test*

The hypotheses for Levene’s test are: 

<i>H<i/><sub>0</sub>: "the population variances of group 1 and 2 are equal"
    
<i>H<i/><sub>1</sub>: "the population variances of group 1 and 2 are not equal"

Burada ise iki grubun varyanslarının eşit olup olmadığına bakıyoruz. Bunu da levene testi ile yapıyoruz. Ve sonuçta p value değeri alpha değerinden büyük çıktığından dolayı H0 hipotezini reddetmekte başarısız oluyorum yani iki grubun varyanslarının eşit olduğunu söyleyebilirim.

```python
stat, p = stats.levene(df[df["gender"]=="Male"]["temperature"],df[df["gender"]=="Female"]["temperature"])

print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
	print('The population variances of group 1 and 2 are equal (fail to reject H0)')
else:
	print('The population variances of group 1 and 2 are not equal (reject H0)')
```

___🚀Conduct the significance test. Use scipy.stats.ttest_ind

H0: µ1 = µ2 ("the two population means are equal")

H1: µ1 ≠ µ2 ("the two population means are not equal")

```python
stats.ttest_ind(df[df["gender"]=="Male"]["temperature"], df[df["gender"]=="Female"]["temperature"], equal_var=True)
```

```python
twosample = stats.ttest_ind(df[df["gender"]=="Male"]["temperature"], df[df["gender"]=="Female"]["temperature"],equal_var=True)

alpha = 0.05
p_value = twosample.pvalue

if p_value<alpha:
    print('At {} level of significance, we can reject the null hypothesis in favor of the alternative hypothesis.'.format(alpha))
else:
    print('At {} level of significance, we fail to reject the null hypothesis.'.format(alpha))
```

Conclusion : Sonuc olarak yukarıda erkek ve kadınların vücut sıcaklıklarının ortalamaları arasında istatiksel olarak anlamlı bir fark var.

Null yani 0 hipotezi her zaman eşitlik üzerine kurulur ve No difference olarak söyleyebiliriz. Alternative hipotez ise eşit değildir üzerine kurulur ve Yes difference olarak söyleyebiliriz. Eğer one tail durumu söz konusu ise alternative hipotezi bu durumda greater right tail veya smaller left tail olarak kurulur.
