# merge

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
display(left,right)
```

```python
pd.merge(left, right)  # default olarak ortak olanı yani key buldu.
```

```python
pd.merge(left, right, on="key")  # ortak sütunlardaki ortak satırları yazdı.
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                     'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']})
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
display(left, right)
```

```python
pd.merge(left, right, on="key", how = "inner") # inner normalde default olandır.
# ortak sütundaki ortak satırları yazıyor.
```

```python
pd.merge(left, right, on="key", how = "outer") # önce key deki değerleri unique olarak yazdı.
```

```python
pd.merge(left, right, on="key", how = "left")
```

```python
pd.merge(left, right, on="key", how = "right")
```

```python
df11 = pd.DataFrame({'lkey': ['x', 'y', 'z', 'x', 'z'],
                    'lvalue': [2, 3, 5, 7, 0]})
df12 = pd.DataFrame({'rkey': ['a', 'x', 'z', 'b'],
                    'rvalue': [7, 8, 9, 10]})
```

```python
display(df11,df12)
```

```python
pd.merge(df11,df12,left_on="lkey",right_on="rkey",how="inner")
```

```python
pd.merge(df11,df12,left_on="lkey",right_on="rkey",how="left")
```

```python
pd.merge(df11,df12,left_on="lkey",right_on="rkey",how="outer")
```

```python
df_left = pd.DataFrame({'lkey': ['x', 'y', 'z', 'x'],
                    'lvalue': [2, 3, 5, 7]})
df_right = pd.DataFrame({'rkey': ['a', 'b', 'c', 'b'],
                    'rvalue': [7, 8, 9, 10]})
```

```python
pd.merge(df_left,df_right,left_on="lkey",right_on="rkey",how="left")
```

```python
display(left4,right4)
```

```python
"".join(["A ","B ","C ","D "])
```

```python
left4.join(right4)  # birini alıp indekslere göre diğerine katıyor.
```

```python
left4.join(right4, how="inner") 
```

```python
left4.join(right4, how="outer") 
```

```python
left4.join(right4, how="right") 
```
