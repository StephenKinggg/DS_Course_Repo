
M(r|s|rs)\.?\s[A-Z]\w*   bu durumda Mr veya Mrs ile başlayıp sonrasında . olan veya olmayan sonrasında ise Büyük harfler ile başlayan ve sonrasında 0 veya daha fazla word karakter olanları getirir.


```python
text = "stephen.kingg@edu.uk"

num = re.search("[a-zA-z]+\.+[a-zA-Z]+@[a-zA-Z]+\.uk", text)   # {} içine yazdığımız kadar karakteri tutar.
print(num)
print(num.group())
```

```python
text = "stephen-321-king@my-work.net"

num = re.search("[a-zA-z0-9.-]+@[a-zA-Z.-]+\.net", text)   # {} içine yazdığımız kadar karakteri tutar.
print(num)
print(num.group())
```

```python
pattern = "[a-zA-Z0-9]+@[a-zA-Z]+\.(com|edu|net)"
user_input= input()
if(re.search(pattern, user_input)):
    print("valid email")
else:
    print("invalid email")
```

```python
pattern = "(\d{3})-(\d{3})-(\d{4})"
new_pattern = r"\1\2\3"
user_input = input()
new_user_input = re.sub(pattern, new_pattern, user_input)
print(new_user_input)
```
