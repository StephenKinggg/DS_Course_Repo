```python
tables
```

```python
pop = tables[5]  # 5. tabloyu aldÄ±k.
```

```python
pop
```

```python
pop.to_sql(name="populations", con=temp_db)  # tabloyu sql atarak nereye baÄŸlayacaÄŸÄ±mÄ± con ile saÄŸlÄ±yorum.
```
