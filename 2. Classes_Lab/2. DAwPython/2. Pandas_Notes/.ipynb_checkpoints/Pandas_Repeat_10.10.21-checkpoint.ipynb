```python
import seaborn as sns
df = sns.load_dataset('diamonds')
df = df.select_dtypes(include = ['float64', 'int64']) 
df = df.dropna()
df
```

```python
sns.boxplot(x=df.table)
```

```python
df_table = df.table
df_table
```

```python
df_table.describe()
```

```python
Q1 = df_table.quantile(0.25)
Q3 = df_table.quantile(0.75)
IQR = Q3-Q1
IQR
```

```python
lower_lim = Q1 - (1.5*IQR)
lower_lim
upper_lim = Q3 + (1.5*IQR)
upper_lim
```

```python
(df_table > upper_lim).value_counts()
(df_table < lower_lim).value_counts()
```

```python
not_outlier = (df_table >= lower_lim) & (df_table <= upper_lim)
not_outlier
```

```python
df_drop_outlier = df[not_outlier]
df_drop_outlier
```

```python
df[ ~((df_table >= lower_lim) & (df_table <= upper_lim))]  # outlier olanları çektik.
```

```python
df_table.loc[ df_table > upper_lim ] = upper_lim
df_table
```

```python
df_table.loc[ df_table < lower_lim ] = lower_lim
df_table
```

```python
sns.boxplot(x = df_table)
```

```python
sns.histplot(x = df_table, kde=True)
```

```python
df.loc[ df.table > lower_lim, "table"] = upper_lim
df
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                     'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']})
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
pd.merge(left, right, on="key")  # ortak sütunlardaki ortak satırları yazdı.
```

```python
pd.merge(left, right, on="key", how = "inner") # inner normalde default olandır.
# ortak sütundaki ortak satırları yazıyor.
```

```python
pd.merge(left, right, on="key", how = "outer") # önce key deki değerleri unique olarak yazdı.
```

```python
pd.merge(left, right, on="key", how = "left")
```

```python
pd.merge(left, right, on="key", how = "right")
```

```python
df11 = pd.DataFrame({'lkey': ['x', 'y', 'z', 'x', 'z'],
                    'lvalue': [2, 3, 5, 7, 0]})
df12 = pd.DataFrame({'rkey': ['a', 'x', 'z', 'b'],
                    'rvalue': [7, 8, 9, 10]})
```

```python
pd.merge(df11,df12,left_on="lkey",right_on="rkey",how="inner")
```

```python
pd.merge(df11,df12,left_on="lkey",right_on="rkey",how="left")
```

```python
pd.merge(df11,df12,left_on="lkey",right_on="rkey",how="outer")
```

```python
df_left = pd.DataFrame({'lkey': ['x', 'y', 'z', 'x'],
                    'lvalue': [2, 3, 5, 7]})
df_right = pd.DataFrame({'rkey': ['a', 'b', 'c', 'b'],
                    'rvalue': [7, 8, 9, 10]})
```

```python
pd.merge(df_left,df_right,left_on="lkey",right_on="rkey",how="left")
```

```python
pd.merge(df_left,df_right,left_on="lkey",right_on="rkey",how="right")
```

```python
pd.merge(df_left,df_right,left_on="lkey",right_on="rkey",how="outer")
```

```python
pd.merge(df_left,df_right,left_index=True, right_index=True, how="outer")
```

```python
df4 = pd.DataFrame({'key': ['K0', 'K2', 'K3', 'K4', 'K5', 'K6'],
                   'X': ['X0', 'X2', 'X3', 'X4', 'X5', 'X6']})
other = pd.DataFrame({'key': ['K0', 'K2', 'K3'],
                      'Y': ['Y0', 'Y2', 'Y3']})
```

```python
df4.join(other, lsuffix="_df", rsuffix="_other")
```

```python
df4.join(other, lsuffix="_df", rsuffix="_other", how="inner")
```

```python
df4.join(other.set_index("key"),on="key")
```
