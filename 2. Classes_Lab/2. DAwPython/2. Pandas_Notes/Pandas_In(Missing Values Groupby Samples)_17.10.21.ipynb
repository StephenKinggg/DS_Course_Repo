```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
new1_df = df.iloc[:,0:3]
new1_df
```

```python
a = new1_df.groupby("Location").sum()
index = a.MinTemp.index
valuemin = a.MinTemp
valuemax = a.MaxTemp
```

```python
plt.plot(index, valuemin)
plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(10,5))

p = np.arange(len(index))  # gün uzunluğu kadar bir array oluşturdum.
width = 0.4  # bir barın genişliği olarak atadığımız değer.

ax.bar(p - width/2, valuemin, width, label="total_bill")  # p-width/2 dersek merkezden bunu bu kadar sola kaydır.
ax.bar(p + width/2, valuemax,        width, label="tip")      # p+width/2 dersek merkezden bunu bu kadar sola kaydır.

ax.set_xticks(p)
ax.set_xticklabels(index)

plt.legend()
plt.show()
```
