```python
planets.groupby(["method","year"]) [["distance","orbital_period"]].std() # grupby parantezinden sonra tek bir 
# sütun ise tek liste daha fazla sütun ise liste icinde göndermeliyiz. Std ile de bunların std sini buluyoruz.
```

```python
df6.groupby("groups").agg({"var1":("mean","median","sum"),
                            "var2":("std","var")
                          })
```

```python
df6.groupby("groups").agg({"var1":(np.mean,np.median),
                            "var2":("std","var")
                          })
```

```python
df6.groupby("groups").filter(lambda x: x["var1"].mean()>19) #filter kendi içindeki koşula göre tablonun ilk haline göre işlem yapıyor.
```

```python
df6[df6["var1"]>20]
```

```python
df6.groupby("groups").std()
```

```python
df6.groupby("groups").filter(lambda x: x["var2"].std()>100)
```

```python
df6[["var1","var2"]].transform(lambda x:(x-x.mean())/x.std())   # z score 
```
