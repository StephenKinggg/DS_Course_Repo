## Train Test Split

## Modelling

```python
model = Sequential()
```

embeddingsize, overfitting ve underfitting kumanda eden parametrelerden biridir. 50, 100, 150, 300 gibi rakamlar kullanılır. İmbalanced datalarda bu parametre overfitting sebep olabiliyor.
BERT modellerde farklı sayılar kullanılabilir.

```python
embedding_size = 50
```

```python
# pip install -U numpy==1.19.5
```

En fazla kullanılan 15000 tokeni seçtik. her bir tokeni 61 sabitledik. ve bunların herbiri 50 vektörel uzunluğunda wordembeddingten oluşuyor. Aşağıdaki Embedding içine yazdıklarım ile 3 boyutlu bir matriks oluşturmuş oldum. num_xords; 15000 token uzunluğunda, output_dim;50 vektörel uzunluğunda, input_length; 61 uzunluğuna sabitlemiştik. Bu şekilde bir embedding layer oluşturmuş olduk. Yani burada modele nasıl bir structure bilgi vereceğimiz söylemiş olduk. eğer bunu yapmasak hata alırız.

Overfitting gitme durumu söz konusu olduğunda 50 lik vektörün içinde en fazla %20 kapatacak şekilde Dropout kullandık. Ancak buradaki 0,2 demek her seferinde aynı sayıda olacak diye bir durum söz konusu değil. En fazla bu rakamı kapatacak bazen 0,1 veya 0,15 de kapatabilir.

units 48 ile burada wordembedding boyutunu düşürmüş oluyoruz. 50 den 48 çekmiş oluyoruz. Biz cell state hidden state eşitliyorduk. ancak bizim son hidden state çok fazla keyword ile dolunca bu şekilde bir sonraki layera verince unutmaları engellemek için return_sequences=True ile de bir sonraki layera aktarılmasını istiyoruz. Yani farklı t zamanlarında elde ettiğimi farklı hidden layerlardaki tüm tokenlerin kaybolmadan bir sonraki layera aktarılmasını sağlamış oluyor.

RNN modellerde art arda oluşturulan layer sayısının en fazla 5 olması tavsiye edilmektedir.

overfitting için embeddingsize, learningrate ve dropout kullandık. Burada batchnormalization kullanmaya gerek yok. Bu gradient exploding engelleyemiyor. 

```python
model.add(Embedding(input_dim=num_words,
                    output_dim=embedding_size,
                    input_length=max_tokens))

model.add(Dropout(0.2))

model.add(GRU(units=48, return_sequences=True))
model.add(Dropout(0.2))

#model.add(GRU(units=24, return_sequences=True))
#model.add(Dropout(0.2))

model.add(GRU(units=12))

model.add(Dense(1, activation='sigmoid'))
```

```python
optimizer = Adam(learning_rate=0.01)
```

```python
model.compile(loss='binary_crossentropy',
              optimizer=optimizer,
              metrics=['Recall'])
```

https://stats.stackexchange.com/questions/328926/how-many-parameters-are-in-a-gated-recurrent-unit-gru-recurrent-neural-network

Aşağıda fit kısmında batchsize belirtmediğimden dolayı nan dolduruyor.

```python
model.summary() 
# 3 x (n2 + nm + 2n) m= input, n= output
# (48*48 + 100*48 +2*48)*3 ==> m = 100, n = 48
# (24*24 + 48*24  +2*24)*3 ==> m =  48, n = 24
# (12*12 + 24*12  +2*12)*3 ==> m =  24, n = 12
```

Aşağıda mode=auto yazarsak monitor yazdığımızda her epochta trendi takip eder eğer artma eğiliminde ise bu max arıyor der ve devreye girer. Ancak buradaki gibi ilk epochta değer max çıkar ve sonra düşmeye başlarsa bu durumda trend azalma eğiliminde der ve min olması gerektiğine karar verir. bu yüzden bunu engellemek için mode=max olarak aldık. 

NLP modellerine recall ve precission takip ederken mode=max, loss ve val_loss takip ederken ise mode=min olarak yazmamızda fayda vardır. NLP de en yüksek skorlar genellikle ilk epochta çıkar.

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(monitor="val_recall", mode="max", 
                           verbose=1, patience = 1, restore_best_weights=True)
```

```python
from sklearn.utils import class_weight
classes_weights = class_weight.compute_sample_weight(class_weight='balanced', y=y_train)
pd.Series(classes_weights).unique()
```

https://keras.io/getting_started/faq/#why-is-my-training-loss-much-higher-than-my-testing-loss

Batchsize sayısı küçüldükçe overfittinge götürme ihtimali daha da artıyor. overfitting ile mücadele ederken bunu kademe kademe artırmak lazım.

Eğer herşeyi denedik ve overfitting engelleyemiyorsak bu durumda tek bir epoch yaparak overfitting engellemeye çalışabiliriz.

loss değeri ne kadar 0 yakınsa iki sınıfı da çok yüksek olasılık ile tahmin ediyor demektir. 


!!!! önemli:

NLP de dengesiz veri setlerinde buradaki loss değerinin ML deki accuracy den farkı yoktur. 

NLP de çok fazla epoch yapmayacağız. Genellikle train değerleri daha yüksek olacak.

```python
model.fit(X_train, y_train, epochs=10, batch_size=256, sample_weight= classes_weights,
         validation_data=(X_test, y_test), callbacks=[early_stop])
```
