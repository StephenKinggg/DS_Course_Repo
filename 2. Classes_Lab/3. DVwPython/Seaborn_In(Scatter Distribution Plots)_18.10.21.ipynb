```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.scatterplot(x= tips.total_bill, y= tips.tip)
```

```python
plt.figure(figsize=(12,8))

ax = sns.scatterplot(x= tips.total_bill, y= tips.tip)   # ax = sns.scatterplot(x= "total_bill", y= "tips", data=tips)

ax.set(xlabel = "Total Bill", ylabel="Tip", title="Some Title")

#plt.legend(labels=["Legend"])

plt.show()
```

***
Hue, style veya size ile kategorik veriyi görselleştiriyoruz.

***

```python
sns.scatterplot(x= "total_bill", y= "tip", data=tips, hue="sex", style="sex", s=200)

plt.show()

# Hue allows to size based on another column.
# s circle ların size nı büyütüyor.
# style, Grouping variable that will produce points with different markers. Can have a numeric dtype but will always be treated as categorical.
```

```python
sns.scatterplot(x= "total_bill", y= "tip", data=tips, size="sex")

plt.show()

# hue koymadığımızda böyle size attığımızda bu şekilde bir grafik çıkıyor. Buda çok tercih edilen bir grafik değil.
```

```python
sns.scatterplot(x='total_bill',y='tip',s=200, data=tips, alpha=1);

#sns.scatterplot(x='salary',y='sales',data=df,s=200,linewidth=0,alpha=0.2)

#Yukarıdaki grafikten farklı olarak alpha ile transparanlık sağlıyor ve nerede yoğunlaşma olduğunu görebiliyoruz.
```

```python
plt.figure(figsize=(12,8))  # görüntüyü büyüttük.
sns.scatterplot(x='total_bill',y='tip', data=tips, style='sex', hue='sex', s=100)

# Call savefig in the same cell
plt.savefig('example_scatter.jpg')
```

```python
sns.rugplot(tips["total_bill"]);
```

```python
sns.rugplot(x="total_bill", data = tips, height=0.025)   # 0.025 default olarak verilen değer.
# height : proportion of axes extent covered by each rug element.
```

```python
sns.displot(x="total_bill", data = tips)   # sns.displot(tips["total_bill"])
```

```python
sns.displot(x="total_bill", data = tips , kde=True)
# gaussian grafiğini barplotların üzerine giydiriyor.
```

```python
sns.histplot(x="total_bill", data = tips , kde=True)
```

```python
sns.histplot(x="total_bill", data = tips, bins=30, kde=True)
# bins ile kutucukların değerleri ile oynuyoruz. Bins artırmak bizim gözlem kalitemizi artırır.
```

```python
sns.set(style='darkgrid')   
sns.histplot(data = tips, x = 'total_bill',bins=20)
```

```python
sns.set(style='darkgrid', rc={"grid.color": ".6", "grid.linestyle": ":"})
sns.histplot(data = tips, x = 'total_bill',bins=20);
```

```python
sns.displot(data = tips, x = 'total_bill',bins=20,kde=False,
            color='red',edgecolor='black',lw=4,ls='--')
```
