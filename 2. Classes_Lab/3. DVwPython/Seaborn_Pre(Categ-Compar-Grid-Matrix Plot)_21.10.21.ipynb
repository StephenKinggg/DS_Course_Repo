## Violin Plot

```python
cars = sns.load_dataset("mpg").dropna()
```

```python
cars.head()
```

```python
cars.describe()
```

```python
sns.set_style("whitegrid")
```

```python
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.violinplot(cars.displacement)
```

```python
sns.violinplot(x="cylinders", y="displacement", 
               data=cars)
```

```python
sns.violinplot(x="cylinders", y="displacement", 
               data=cars, hue="origin"
               )
```

```python
sns.violinplot(x="cylinders", y="displacement", 
               data=cars[cars.origin.isin(["japan","europe"])], hue="origin"
               )
plt.legend(loc=2)
```

```python
sns.violinplot(x="cylinders", y="displacement", 
               data=cars[cars.origin.isin(["japan","europe"])], hue="origin",
               split=True
               )

plt.legend(loc=2)   
# legend location belirliyor. 2(üst sol)  1(üst sağ)/4(alt sol) 3(alt sol) köşelere göre. 
```

```python
sns.violinplot(x="cylinders", y="displacement", 
               data=cars[cars.origin.isin(["japan","europe"])], hue="origin",
               split=True,
               inner="quartiles"
               )

plt.legend(loc=2) 
```

```python
sns.violinplot(x="cylinders", y="displacement", 
               data=cars[cars.origin.isin(["japan","europe"])], hue="origin",
               split=True,
               scale="count"
               )
# scale : {"area", "count", "width"}, optional
# The method used to scale the width of each violin. If ``area``, each violin will have the same area. If ``count``, the width of th
plt.legend(loc=2)
```

```python
sns.violinplot(x="cylinders", y="displacement", 
               data=cars[cars.origin.isin(["japan","europe"])], hue="origin",
               split=True,
               scale="count",
               scale_hue=False
               )
# scale : {"area", "count", "width"}, optional
# The method used to scale the width of each violin. If ``area``, each violin will have the same area. If ``count``, the width of th
plt.legend(loc=2)
```

```python
sns.violinplot(x=cars.origin, y=cars.displacement,
              order=(["japan","europe","usa"])
              )
```

```python
sns.violinplot(x=cars.origin, y=cars.displacement,
              order=(["japan","europe","usa"]),
               linewidth=4   # çerçeve çizginin kalınlığı.
              )
```

***
kde options
***

```python
sns.violinplot(x=cars.origin, y=cars.displacement,
              order=(["japan","europe","usa"]),
              bw=0.4  # çerçeve çizginin kalınlığı.
              )
# bw : {'scott', 'silverman', float}, optional
# Either the name of a reference rule or the scale factor to use when computing the kernel bandwidth. The actual kernel size will be
# determined by multiplying the scale factor by the standard deviation of the data within each bin.
```

## Joint Plot

```python
sns.jointplot(x="waiting", y="duration", data=geyser)
```

***
There are currently six kinds of jointplots allowed in seaborn:
   1. scatter(default)
    
   2. kde
    
   3. reg
    
   4. hist
   
   5. hex
    
   6. resid
***

```python
sns.jointplot(x="waiting", y="duration", data=geyser,
             kind="scatter")
```

```python
sns.jointplot(x="waiting", y="duration", data=geyser,
             kind="kde")
```

```python
sns.jointplot(x="waiting", y="duration", data=geyser,
             kind="reg")
```

```python
sns.jointplot(x="waiting", y="duration", data=geyser,
             kind="hist")
```

```python
g = sns.jointplot(x="waiting", y="duration", data=geyser)
```

```python
type(g)
```

```python
g.plot_joint(sns.kdeplot, color="gray", levels=5)
```

```python
sns.jointplot(x="waiting", y="duration", data=geyser,
             hue="kind")
```

```python
sns.jointplot(x="waiting", y="duration", data=geyser,
             hue="kind",
             kind="kde")
```

```python
sns.jointplot(x="waiting", y="duration", data=geyser,
             hue="kind",
             palette="autumn")
```

```python
sns.jointplot(x="waiting", y="duration", data=geyser,
              height=6,
              space=0.2,
              ratio=2     #ratio 5 default.
             )
```
