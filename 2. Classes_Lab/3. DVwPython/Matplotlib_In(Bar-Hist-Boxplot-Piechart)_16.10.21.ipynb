***
Scatter plot , değişkenler arasındaki ilişkiyi ifade etmek için,

Barplot; nominal ve ordinal yani kategorik değişkenler içeren veri setlerinin analizinde,

Piecharts; nominal ve ordinal yani kategorik değişkenler içeren veri setlerinin analizinde,

Histogram; intervel ve ratio yani numerik değişkenler içeren veri setlerinin analizinde,

Boxplot; intervel ve ratio yani numerik değişkenler içeren veri setlerinin analizinde kullanılır.
***

```python
day = tips.day.unique()
day
```

```python
tips.groupby("day")["total_bill"].sum()
```

```python
day_of_total_bill= list(tips.groupby("day")["total_bill"].sum())
day_of_total_bill
```

```python
fig, ax = plt.subplots(figsize = (12,7))  # figsize ile ölçülerini genişlettik.

ax.barh(day, day_of_total_bill)
plt.show()
```

```python
day_of_tip= list(tips.groupby("day")["tip"].sum())
day_of_tip
```

```python
fig, ax = plt.subplots(figsize = (8,6))  # figsize ile ölçülerini genişlettik.

ax.bar(day, day_of_tip)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,6))  # figsize ile ölçülerini genişlettik.


ax.bar(day, day_of_total_bill, label="total_bil")
ax.bar(day, day_of_tip, label="tip")

plt.legend()

plt.show()
```

```python
# Optional hard to coding
fig, ax = plt.subplots(figsize=(10,5))

p = np.arange(len(day))  # gün uzunluğu kadar bir array oluşturdum.
width = 0.4  # bir barın genişliği olarak atadığımız değer.

ax.bar(p - width/2, day_of_total_bill, width, label="total_bill")  # p-width/2 dersek merkezden bunu bu kadar sola kaydır.
ax.bar(p + width/2, day_of_tip,        width, label="tip")      # p+width/2 dersek merkezden bunu bu kadar sola kaydır.

ax.set_xticks(p)
ax.set_xticklabels(day)

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,8))

y = np.array([35, 25, 25, 15])

mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

myexplode = [0.1, 0, 0, 0]
```

```python
plt.figure(figsize=(10,8))
plt.pie(y, labels = mylabels, explode = myexplode, labeldistance = 0.7, autopct="%.2f")
plt.show()
```

```python
plt.figure(figsize=(10,8))

colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0, 0, 0.1, 0.1] # ratio for each continent with which to offset each wedge.

y = np.array([25, 35, 25, 15])
mylabels = ["Bananas", "Apples", "Cherries", "Dates"]
myexplode = [0.1, 0.1, 0, 0]

plt.pie(y, labels = mylabels, labeldistance=0.5, autopct="%.1f", startangle=90, shadow=True, pctdistance=1.1, colors=colors_list, explode=explode_list)
plt.legend()
plt.show()

# startangle ilk olan Bananyı 90 derece olarak önce onu yerleştiriyor sonra diğerlerini de ona göre yerleştiriyor.
# autopct bunları float olarak nokta sonrasına koyduğumuz kadar basamak koyar. 
# labeldistance labelları içeri ve dışarı doğru
```
