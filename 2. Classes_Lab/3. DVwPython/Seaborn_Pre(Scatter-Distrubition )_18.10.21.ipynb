***
kind : {"hist", "kde", "ecdf"}

kde = kernel density estimation

ecdf = empirical distribution functional plot
    
Approach for visualizing the data. Selects the underlying plotting function and determines the additional set of valid parameters.
***

```python
sns.displot(x = "weight", data = cars, kind="ecdf")
```

```python
sns.displot(x = "weight", data = cars, kind="kde", rug= True)

# rug : bool
# If True, show each observation with marginal ticks (as in :func:`rugplot`).
# alt taraftaki yukarÄ± doÄŸru olan Ã§izgiler.
```

```python
sns.displot(x = "weight", data = cars, hue="origin")

# hue : categorical variables.
# hue : vector or key in ``data``
# Semantic variable that is mapped to determine the color of plot elements.
```

```python
# Styling
## Palette
```

***
palette : string, list, dict, or :class:`matplotlib.colors.Colormap`

Method for choosing the colors to use when mapping the ``hue`` semantic.String values are passed to :func:`color_palette`. List or dict values imply categorical mapping, while a colormap object implies numeric mapping.
***

```python
sns.displot(x ="weight", data = cars,
           col = "origin", hue = "origin",
           palette = "crest")
```

```python
# height, aspect
```

***
height : scalar
    Height (in inches) of each facet. See also: ``aspect``.    
aspect : scalar
    Aspect ratio of each facet, so that ``aspect * height`` gives the width
***

```python
sns.displot(x ="weight", data = cars,
           col = "origin", height = 5, aspect=0.5)

# aspect width/height olduÄŸundan dolayÄ± aspect artarsa height kÃ¼Ã§Ã¼lÃ¼r.
```

```python
sns.displot(x ="weight", data = cars,
           bins=10)

# bins bÃ¼yÃ¼dÃ¼kÃ§e kutucuklarÄ±n geniÅŸlik kÃ¼Ã§Ã¼lÃ¼r ve grafiÄŸi daha iyi okumamÄ±zÄ± saÄŸlar.
```

```python
sns.displot(x ="weight", data = cars,
           kind = "kde", fill =True, lw=3)

# fill ile grafiÄŸin alt tarafÄ±nÄ± tarÄ±yoruz. 
# lw ise grafiÄŸin dÄ±ÅŸ cephesini kalÄ±nlaÅŸtÄ±rÄ±yor.
```

```python
sns.displot(x ="weight", data = cars,
           kind = "kde", rug =True, rug_kws = {"height" : 0.15})

# rug_kws ile grafiÄŸin altÄ±ndaki ruglarÄ±n yÃ¼ksekliÄŸini belirliyoruz. 0.025 default olarak atanÄ±yor.
```
