```python
days=[1,2,3,4,5,6,7]
max_t=[32,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel("Days")
plt.xlabel("Temparature")
plt.xlabel("Weather")

plt.plot(days,max_t, label="Max")
plt.plot(days,min_t, label="Min")
plt.plot(days,avg_t, label="Avg")

plt.legend(loc="best", shadow=True,fontsize="small")  
# label i verilerin durumuna göre en uygun boşluğa alır. Farklı locationla var.

plt.grid()
```

```python
import numpy as np
x = np.linspace(0,10,10)
y1 = x
y2 = x**2
y3 = x**3
y4 = np.sqrt(x)

plt.figure()
plt.subplot(2,2,1)
plt.plot(x,y1,"ro")
plt.title("$y_1=x")

plt.subplot(2,2,2)
plt.plot(x,y2,"g--")
plt.title("$y_2=x^2$")

plt.subplot(2,2,3)
plt.plot(x,y3,"b^")
plt.title("$y_3=x^3$")

plt.subplot(2,2,4)
plt.plot(x,y1,"ks")
plt.title("$y_4=\sqrt{x}$")
```

```python
x = np.arange(0,30,0.2)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show
```

```python
plt.title("Sine Function")
plt.suptitle("I am bigger")

plt.plot(x,y)
plt.show
```

```python
plt.title("Sine Function")
plt.xlabel("Weight of Students")
plt.ylabel("Height of Students")

plt.plot(x,y)
plt.show
```

```python
x = np.arange(0,30,0.2)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.title("Sine Function")
plt.xlabel("Weight of Students")
plt.ylabel("Height of Students")

plt.plot(x,y1, label="Sine")
plt.plot(x,y2, label="Cosine")
plt.legend(loc= "upper left")

plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
company = ["GOOGL","AMZN","MSFT","FB"]
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

```python
ypos = np.arange(len(company))
ypos
```

```python
plt.bar(ypos, revenue)
```

```python
plt.xticks(ypos,company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stoks")
plt.bar(ypos, revenue, label="Revenue")
plt.bar(ypos,profit,label="Profit")
plt.legend()
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(ypos,company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stoks")
plt.bar(xpos-0.2, revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2, profit, width=0.4, label="Profit")
plt.legend()
```

```python
plt.xticks(ypos,company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stoks")
plt.bar(xpos-0.2, revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2, profit, width=0.2, label="Profit")
plt.legend()
```

```python
plt.xticks(ypos,company)
plt.title("US Tech Stoks")


plt.barh(xpos-0.2, revenue, label="Revenue")
plt.barh(xpos+0.2, profit, label="Profit")
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar)
```

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar, bins=3)
```

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar, bins=3, rwidth=0.95)
```

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar, bins=[80,100,120,160] ,rwidth=0.95, color= "g", histtype="step")
```

```python
blood_sugar_men= [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women= [67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.xlabel("sugar range")
plt.ylabel("Total no of patients")
plt.title("Blood sugar analysis")

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,120,160], rwidth=0.95, color= ["green","orange"], 
         label=["men","women"], orientation="horizontal")
plt.legend()
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill","Car","Other Utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels = exp_labels, radius=1.5)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels = exp_labels, radius=1.5,autopct ="%0.1f%%", shadow=True)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels = exp_labels, radius=1.5,autopct ="%0.1f%%", shadow=True, explode=[0,0.1,0.1,0.1,0.1], startangle=180)
plt.show()

#explode seçili dilimleri ayırıyor. startangle istenildiği kadar pie döndürür.
```
