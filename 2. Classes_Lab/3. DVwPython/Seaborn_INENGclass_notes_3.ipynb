# Explore Data

```python
# extended describe()
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import scipy

d = {'a': 123}
d
```

```python
sns.get_dataset_names()
penguins = sns.load_dataset('penguins')
penguins.head(5)
```

```python
penguins.describe(include='all')
```

```python
sns.heatmap(penguins.isnull())
#sns.heatmap(penguins[:20].isnull())
```

# Catplot

```python
sns.catplot(x='species' , y='bill_length_mm' , data=penguins)
```

```python
sns.catplot(x='species', y='bill_length_mm', data=penguins, 
           kind='bar', hue='sex')
```

```python
sns.catplot(x='species', y='bill_length_mm', data=penguins,
            kind='bar', hue='sex', order=['Gentoo', 'Chinstrap', 'Adelie']
           )
```

```python
sns.catplot(x='species', y='bill_length_mm', data=penguins,
            kind='bar', hue='sex', order=['Gentoo', 'Chinstrap', 'Adelie'],
            ci='sd')
```

```python
sns.catplot(x='species', y='bill_length_mm', data=penguins,
            kind='bar', hue='sex', col='island')
```

```python
sns.catplot(x='species', y='bill_length_mm', data=penguins,
            kind='bar', hue='sex', col='island', row= 'sex'
           )
# kind=   'swarm' (swarmplot), violin, boxen, box
```

```python
# Combine different plots
g = sns.catplot(x='species', y='body_mass_g', kind='violin', inner=None, data=penguins)
# Figure(implicit), <target> ~~ matplotlib.pyplot (plt)  ~~plt.plot(); plt.plot()
sns.swarmplot(x='species', y='body_mass_g', data=penguins, color='black', size=3, ax=g.ax)
#bw=
```

# Jointplot

```python
sns.jointplot(x='flipper_length_mm', y='body_mass_g', data=penguins)
```

```python
sns.jointplot(x='flipper_length_mm', y='body_mass_g', data=penguins,
             kind='reg', ci=90)
#kind= 'scatter, kde, hist, hex, reg, resid'
```

```python
sns.jointplot(x='flipper_length_mm', y='body_mass_g', data=penguins,
              kind='kde')
```

```python
sns.jointplot(x='flipper_length_mm', y='body_mass_g', data=penguins,
              kind='hex')
```

```python
sns.jointplot(x='flipper_length_mm', y='body_mass_g', data=penguins,
              hue='species')
```

```python
sns.jointplot(x='flipper_length_mm', y='body_mass_g', data=penguins,
              hue='species', kind='kde')
```

# Pairplot

```python
sns.pairplot(penguins)
```

```python
sns.pairplot(penguins,
             hue='species'
            )
```

```python
sns.set_palette( ['red', 'blue', 'green']    )
sns.pairplot(penguins, hue='species', )
```

```python
# bar, swarm, violin .....
sns.pairplot(penguins, hue='species', corner=True, diag_kind='hist')
```

# Pairgrid

```python
g = sns.PairGrid(penguins)
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(penguins)
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot)
```

```python
g = sns.PairGrid(penguins)
g.map_upper(sns.scatterplot, color='orange')
g.map_lower(sns.kdeplot, color='green')
g.map_diag(sns.histplot)
```

```python
sns.set_palette(['g', 'purple', 'red'])
g = sns.PairGrid(penguins, hue='species')
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot)
g.add_legend()
```

# FacetGrid

```python
g = sns.FacetGrid(data=penguins, col='species',   row='sex')
g.map( plt.hist, "body_mass_g", color='red' )
```
