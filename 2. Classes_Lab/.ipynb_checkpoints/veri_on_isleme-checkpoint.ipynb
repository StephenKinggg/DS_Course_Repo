## Aykırı Değerleri Yakalamak

```python
Machine Learning te kullanacağımız veri ne kadar temiz olursa modelimiz de o kadar düzgün çalışır.
```

Aykırılığı ifade eden nümerik değere aykırı değer denir. 

Veride genel eğilimin oldukça dışına çıkan ya da diğer gözlemlerden oldukça farklı olan gözlemlere aykırı gözlem denir. Yani aykırı değeri barındıran gözlem birimine aykırı gözlem denir.

En önemli problemlerden biridir.

Yukarıda ikinci grafikte yukarıda bulunan 3 nokta grafiğin yönünü değiştirmiş.

## Çok Değişkenli Aykırı Gözlem Analizi

### Local Outlier Factor

Gözlemleri bulundukları konumda yoğunluk tabanlı skorlayarak buna göre aykırı değer olabilecek değerleri tanımlayabilmemize imkan sağlıyor.

Bir noktanın local yoğunluğu bu noktanın komşuları ile karşılaştırılıyor. Eğer bir nokta komşularınının yoğunluğundan anlamlı şekilde düşük ise bu nokta komşularından daha seyrek bir bölgede bulunuyordur yorumu yapılabiliyor. Dolayısıyla burada bir komşuluk yapısı söz konusu. Bir değerin çevresi yoğun değilse demek ki bu değer aykırı değerdir şeklinde değerlendiriliyor.

```python
import seaborn as sns
diamonds = sns.load_dataset('diamonds')
diamonds = diamonds.select_dtypes(include = ['float64', 'int64']) 
df = diamonds.copy()
df = df.dropna()
df.head()
```

```python
import numpy as np
from sklearn.neighbors import LocalOutlierFactor
```

```python
clf = LocalOutlierFactor(n_neighbors = 20, contamination = 0.1)
```

```python
clf.fit_predict(df)
```

```python
df_scores = clf.negative_outlier_factor_
```

```python
df_scores[0:10]
```

```python
np.sort(df_scores)[0:20]
```

```python
esik_deger = np.sort(df_scores)[13]
```

```python
aykiri_tf = df_scores > esik_deger
```

```python
aykiri_tf
```

```python
### silme yöntemi
```

```python
yeni_df  = df[df_scores > esik_deger]
```

```python
yeni_df
```

```python
df[df_scores < esik_deger]
```

```python
df[df_scores == esik_deger]
```

```python
### baskilama
```

```python
baski_deger = df[df_scores == esik_deger]
```

```python
aykirilar = df[~aykiri_tf]
```

```python
aykirilar
```

```python
aykirilar.to_records(index = False)
```

```python
res = aykirilar.to_records(index = False)
```

```python
res[:] = baski_deger.to_records(index = False)
```

```python
res
```

```python
df[~aykiri_tf]
```

```python
import pandas as pd
df[~aykiri_tf] = pd.DataFrame(res, index = df[~aykiri_tf].index)
```

```python
df[~aykiri_tf]
```

## Kategorik Değişken Kırılımında Değer Atama

```python
V1 = np.array([1,3,6,np.NaN,7,1,np.NaN,9,15])
V2 = np.array([7,np.NaN,5,8,12,np.NaN,np.NaN,2,3])
V3 = np.array([np.NaN,12,5,6,14,7,np.NaN,2,31])
V4 = np.array(["IT","IT","IK","IK","IK","IK","IK","IT","IT"])

df = pd.DataFrame(
        {"maas" : V1,
         "V2" : V2,
         "V3" : V3,
        "departman" : V4}        
)

df
```

```python
df.groupby("departman")["maas"].mean()
```

```python
df["maas"].fillna(df.groupby("departman")["maas"].transform("mean"))
```
