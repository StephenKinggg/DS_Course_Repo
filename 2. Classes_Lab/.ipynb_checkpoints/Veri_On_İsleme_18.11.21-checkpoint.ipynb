## Çok Değişkenli Aykırı Gözlem Analizi

### Local Outlier Factor

Gözlemleri bulundukları konumda yoğunluk tabanlı skorlayarak buna göre aykırı değer olabilecek değerleri tanımlayabilmemize imkan sağlıyor.

Bir noktanın local yoğunluğu bu noktanın komşuları ile karşılaştırılıyor. Eğer bir nokta komşularınının yoğunluğundan anlamlı şekilde düşük ise bu nokta komşularından daha seyrek bir bölgede bulunuyordur yorumu yapılabiliyor. Dolayısıyla burada bir komşuluk yapısı söz konusu. Bir değerin çevresi yoğun değilse demek ki bu değer aykırı değerdir şeklinde değerlendiriliyor.

Yukarıdaki örnekte olduğu gibi değişken tek başına olduğunda değerler aykırı değer oluşturmayabilir ancak başka değişken ile birlikte değerlendirildğinde ise aykırı bir değer söz konusu olabilir.

Aşağıdaki örnekte sağdaki değerlerin komşuları ile yoğunluk söz konusu. A da ise diğerleri ile kıyaslandığında çok düşük bir yoğunluk söz konusu.

```python
import seaborn as sns
diamonds = sns.load_dataset('diamonds')
diamonds = diamonds.select_dtypes(include = ['float64', 'int64']) 
df = diamonds.copy()
df = df.dropna()
df.head()
```

```python
import numpy as np
from sklearn.neighbors import LocalOutlierFactor
```

```python
clf = LocalOutlierFactor(n_neighbors = 20, contamination = 0.1)  
# n komşuluk sayısını ifade eden bir argümandır. contamination ise yoğunluğu ifade ediyor.
```

```python
clf.fit_predict(df)  # herbir gözlem birimi için elde edilmiş score lar söz konusu.
```

```python
df_scores = clf.negative_outlier_factor_
```

```python
df_scores[0:10]
```

```python
np.sort(df_scores[0:10])
```

```python
np.sort(df_scores[0:20])
```

```python
np.sort(df_scores)[0:20] 
#Yukarıdaki işlem ilk 20 değeri sıralıyor bu isi tamamını sıralayıp ilk 20 ni alıyor.
# bu işleme bir eşik değer belirleyip ona göre devam etmek lazım.
```

```python
esik_deger = np.sort(df_scores)[13]  # Burada 13 rastgele aldık. Bunu değişimin olduğu değere bakıpta alabilirdik.
#yukarıdaki sıralamaya göre 13.değeri aldık. ve bunu eşik değer olarak atadık.
```

```python
aykiri_tf = df_scores > esik_deger #eşik değerden büyük olanları aldık.
```

```python
aykiri_tf
```

```python
### silme yöntemi
```

```python
yeni_df  = df[df_scores > esik_deger]  # aykırı olmayan değerleri aldık.
```

```python
yeni_df   # eşik değerin üzerinde olan tüm değerleri aldık. Yani aykırı olmayan değerlere ulaştık.
```

```python
df[df_scores < esik_deger]  

#bunlar ise aykırı olan değerler. çok değişkenlide aykırı değer olarak gelen değer sayısı tek değişkenlidekine göre daha az olduğu dikkatimizi çekti.
```

```python
df[df_scores == esik_deger]  # eşik değere eşit olan indeksi getirdik. Yani bir gözlem birimi bulduk.
```

```python
### baskilama
```

```python
baski_deger = df[df_scores == esik_deger]  # elimizde bir baskılama için eşik değeri var.
```

```python
aykirilar = df[~aykiri_tf]  # aykırı olmayan değerlerin dışındakileri aykırı değer olarak değerlendirdik.
```

```python
aykirilar   # aykırılıkların yerine baskı değerini atayacağız. Ancak indeks problemleri var onları düzeltmemiz gerekir.
```

```python
aykirilar.to_records(index = False) # df içindeki değerleri alıp bir numpy array çevirdik.
```

```python
res = aykirilar.to_records(index = False)
```

```python
res[:] = baski_deger.to_records(index = False) 

# içindeki tüm değerlere aykırı değerlerin yerine baskı değerlerini yani eşik değerini yerleştirdik.
```

```python
res
```

```python
df[~aykiri_tf] 
# Ancak bunlar ayrı df ler ve ana df içinde bunlar yok. Yukarıdaki array bu df içine yerleştirmemiz gerekir.
```

```python
import pandas as pd
df[~aykiri_tf] = pd.DataFrame(res, index = df[~aykiri_tf].index) 
# index değeri olarak bu df in yani aykırı olanların index numarasını vermesini istiyoruz.
```

```python
df[~aykiri_tf]
```
