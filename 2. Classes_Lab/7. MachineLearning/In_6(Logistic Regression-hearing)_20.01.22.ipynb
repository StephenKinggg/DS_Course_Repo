## Train | Test Split and Scaling

```python
X = df.drop('test_result', axis=1)
y = df['test_result']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)

# Amacımız train seti için olabildiğince veri sağlamak olduğundan dolayı test setine % 10 verdik.
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)  # Data leakage için sadece train setindeki datalara uyguladım.
X_test_scaled = scaler.transform(X_test)
```

## Modelling

```python
from sklearn.linear_model import LogisticRegression
```

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html

```python
# help(LogisticRegression)
```

log.reg. aslında bir linear reg.dır. Bizim linear reg.ile alamadığımız değerleri yani -sonsuz iile +sonsuz arasındaki değerleri log ile arka planda olasılıklara dönüştürüyor. Default değere göre ise bunları 0.5 üstünü 1, altını ise 0 classına ayırır.

log.reg. penalty için default değer l2 dir. Default değerlere çok fazla dokunmak istemiyoruz. Ancak burada log reg üzerinden feature selection yapmak için penalty l1 çevirmemiz gerekir. none yazarsak linear reg. gibi davranır. overfitting çok yatkın bir yöntemdir. Bu istediğimiz bir durum değil. 

log.reg. içinde bir regularization işlemi yapılıyor ve bunu belirleyende C katsayısıdır. alpha değeri büyüdükçe uyguladığı regularization değeri büyüyordu. Ancak alpha nın aksine C katsayısı küçüldükçe regularization değeri büyür.

linear reg.olduğu gibi multicollinearity ve feature selection sorunları burada yok. Çünkü arka planda bunları yapan default olarak yapan bir parametrem var.

```python
log_model = LogisticRegression() # default penalty=l2
```

```python
log_model.fit(X_train_scaled, y_train)  # scale edilmiş datayı kullanıyoruz.
```

```python
log_model.coef_
```

```python
log_model.intercept_  # eksi olursa 0,5 ten küçük olduğunu ve  testten geçemediğini gösterir.

# artı olursa 0,5 ten büyük olduğunu ve testten geçtiğini gösterir.
```

```python
df.head()
```

```python
y_pred = log_model.predict(X_test_scaled) # train sonrası test sonucları üzerinden predict yapıyoruz.
y_pred
```

## ROC (Receiver Operating Curve) and AUC (Area Under Curve)

Purpose of ROC Curve:
    1. Analysing the strength/predictive power of a classifier.
    2. Determining optimal threshold.
    3. Comparing two models (using Area Under the Curve).

AUC represents a degree or measure of separability. It tells us how much the model is capable of distinguishing between classes. Higher the AUC, better the model is at predicting the probability of class YES higher than the probability of class NO.

The ROC Curve was first used during World War II for the analysis of radar signals. After the attack on Pearl Harbor, the US army began new research to improve the rate of detection of Japanese aircraft from their radar signals. Needless to say, they did not want to miss any of them. Neither did they wish to waste their resources on false alarms. They measured the ability of a radar receiver operator to make these predictions called the Receiver Operating Characteristic. That is the origin of the name. The purpose of the curve was similar to how we use it to improve our machine learning models now. The aim was to analyse the predictive power of the predictor in ensuring the detection of as many true positives as possible while minimizing false positives.

https://towardsdatascience.com/demystifying-roc-curves-df809474529a

Burada düşman olanları doğru tespit etmek TP, dost olanları ise yanlış tespit etmek yani dosta düşman demek FP. TP ile FP değerlerini y ekseninde işaretliyoruz. Öyle bir değerlendirme yapmalıyım ki dost ve düşmanı çok iyi ayırt etmem gerekir. 0 ile 1 arasında diyelim ki 100-150 tane herbir threshold için bulduğumuz değerleri plotlarız ve altında kalan alan ne kadar büyükse benim modelim dost ve düşmanı ayırmakta başarılı ve aynı zamanda iyi tahminler yapıyor. Yani düşmanları doğru max seviyede tahmin edeceğiz ve aynı zamanda dost olanları da düşman olarak min seviyede tahmin etmeyeceğiz. Yani en iyi nokta olan sağ üst köşeye ne kadar yaklaşık olarak bu eğrinin oluşmasıdır.
