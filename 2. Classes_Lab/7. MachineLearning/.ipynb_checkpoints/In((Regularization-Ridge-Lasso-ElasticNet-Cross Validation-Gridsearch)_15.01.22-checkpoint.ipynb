# Regularization(Ridge, Lasso, ElasticNet) - Cross Validation - Gridsearch

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = (10,6)
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv("Advertising.csv")
```

```python
df.head()
```

```python
X = df.drop("sales", axis =1)
y = df["sales"]
```

### Polynomial Conversion

```python
from sklearn.preprocessing import PolynomialFeatures
```

We create an artificial overfiting situation by taking poly feature from 5 degrees

```python
polynomial_converter = PolynomialFeatures(degree=5, include_bias=False)
# daha öncesinde degress i 5 olarak bulduğumuzdan dolayı 5 olarak yazdık. 
poly_features = polynomial_converter.fit_transform(X)
```

```python
poly_features.shape
```

### Train | Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
# burada artık yeni feature ları kullanmaya başladık.
```

## Scaling the Data

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

scaler = StandardScaler()
```

```python
scaler.fit(X_train)  

# tum datanın değil sadece Xtrain mean ve std sini alıp bunlardan çhange yapıyor.

# önce train üzerinde uygulanır. Buradaki fit train yapmıyor sadece dönüşümde kullanacağı Xtrain bilgilerini alıyor.
```

```python
X_train_scaled = scaler.transform(X_train)  
# burada ise dönüşüm işlemini yapıyorum. tüm dataya x-mean/sdt yi uyguluyor. Burada Xtest ilişkin herhangi bir bilgi yok.
X_train_scaled   # elimizde z score lardan oluşan yeni bir X train oluşmuş oldu.
```

```python
X_test_scaled = scaler.transform(X_test) 
# burada Xtest ilişkin dönüşüm ise Xtrain üzerinden yapılıyor. 
# Böylece data leakage engellenmiş oluyor.
# burada train yok sadece Xtrain üzerinden yapılan scalere Xtest dönüştürülüyor.
X_test_scaled # elimizde z score lardan oluşan yeni bir X test oluşmuş oldu.
```

### checking std = 1 and mean = 0

this gives us the z-scores. so it's also called z-score 

These values show where in the normal distribution they correspond to the z score.


```python
pd.DataFrame(X_train_scaled).agg(["mean", "std"]).round() 

# round demessek çok küçük sayılar olduğundan uzun çıkıyor.
# burada - ler eksi taraftan + lar ise artı taraftan 0 yakınsamış demek.
```

```python
pd.DataFrame(X_test_scaled).agg(["mean", "std"]).round()
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
```

```python
lm.fit(X_train_scaled, y_train) # targete kesinlikle scale uygulamıyoruz.Çok önemli!!!!!

# içine scale edilmiş datayı kullanıyorum yani Xtrain kullanmıyorum.
```

```python
y_pred = lm.predict(X_test_scaled)  
# burayada scale edilmiş datayı vermemiz gerekir. Yoksa hata alırız.
y_train_pred = lm.predict(X_train_scaled)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def train_val(y_train, y_train_pred, y_test, y_pred, i):
    
    scores = {i+"_train": {"R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),                          
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
    
    i+"_test": {"R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

```python
ls =train_val(y_train, y_train_pred, y_test, y_pred, "linear")
ls  
```

## Multicolineratiy

If there is a strong correlation between the independent variables, this situation is called **multicolineraty**.
 
**Multicolineraty** prevents my model from detecting important features.

```python
def color_red(val):
    
    if val > 0.90 and val < 0.99:  # val. bu aralıkta ise rengini kırmızı değilse siyah bırakıyor.
        color = 'red'
    else:
        color = 'white'
    return f'color: {color}'
```

```python
pd.DataFrame(poly_features).corr().style.applymap(color_red)  # yukarıdaki func buraya uyguladım.
```

cross validation is only applied to the train set.

## Choosing best alpha value with Cross-Validation

```python
from sklearn.linear_model import RidgeCV
```

```python
alpha_space = np.linspace(0.01, 1, 100) 
alpha_space
# burada bir alpha aralığı belirledik. alphaya 0 vermedik çünkü linear reg oluşur ve buda zaten benim overfitting sorunumun sebebidir.
# buradaki değere 1 değilde daha büyük değerler verebiliriz.
# CV sonucu aldığım en iyi alpha değeri  bu aralık sınırlarında çıkarsa,
# aralığı tekrar değiştiririz.
# CV sonucu alpha aralıklarını görerek bunları yeniden kendimiz belirleyebiliriz yukarısı için.
```

```python
ridge_cv_model = RidgeCV(alphas=alpha_space, cv = 5, scoring= "neg_root_mean_squared_error")

# alpha değerini default olarak değil de kendimiz verdik.
# rmse daha önemli olduğundan dolayı bunu verdik. Çünkü tek bir metric verebiliyoruz.
```

```python
ridge_cv_model.fit(X_train_scaled, y_train)

# bütün alpha değerleri için CV yapıyor ve en iyi en düşük rmse değerini hangi alpha 
# değerinde almışsak bunu en iyi alpha değeri olarak alıyor ve bunu ridgecvmodel atıyor.
# artık bu bir model gibi davranıyor.
```

```python
ridge_cv_model.alpha_   # Ridge(alpha=0.02)
# CV aldığı 5 score ortalamasını bana döndürdüğünden dolayı tekrar CV yapmaya gerek yok.
```

```python
ridge_cv_model.best_score_  # en iyi alpha parametresinde aldığı rmse değerini bana döndürdü.
# CV gibi 4 score yerine sadece rmse değerini döndürdü.(Bunu seçmiştik yukarıda)
```

```python
y_pred = ridge_cv_model.predict(X_test_scaled)
y_train_pred = ridge_cv_model.predict(X_train_scaled)
```

```python
rcs = train_val(y_train, y_train_pred, y_test, y_pred, "ridge_cv")
rcs
# görüldüğü gibi ridgecvmodel sonucu aldığım en iyi rmse score olan 0.65 ile 
# ridgecvtest rmse score olan 0.67 birbirine çok yakın olduğunu görüyorum.
```

```python
pd.concat([ls, rs, rcs], axis = 1) # görüldüğü gibi scorelarımız biraz daha iyileşmiş.
```

```python
ridge_cv_model.coef_
```

```python
rcm_df = pd.DataFrame(ridge_cv_model.coef_, columns=["ridge_cv_coef_0.02"])
```

```python
pd.concat([lm_df,rm_df, rcm_df], axis = 1)  # katsayılarımızı karşılaştırıyoruz.
# alpha değeri büyüdükçe katsayıların küçüldüğünü görüyoruz. 
# 0 yakın olmayanları küçültmüş, 0 yakın olanları ise daha da küçültmüş. 
# Agresif davranmış diyebiliriz.
```

## Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
elastic_model = ElasticNet(max_iter=10000, random_state=42) # bunlar mesela score iyileştirecek parametreler değil. Ancak hyper parameter bunlar.
```

```python
param_grid = {"alpha":[0.01, 0.012, 0.2, 0.5, 0.6, 0.7, 1],
            "l1_ratio":[0.1, 0.5, 0.7, 0.9, 0.95, 0.99, 1]}
```

```python
grid_model = GridSearchCV(estimator = elastic_model, param_grid = param_grid, scoring = 'neg_root_mean_squared_error',
                         cv =5, verbose =2)

# param_grip hyper_parametrelerin hangi aralıklarda olmasını istiyorsak manuel olarak yazıyoruz.
# param_grip içine dict olarak veriyoruz yoksa hata verir. alpha ve l1_ration parametreleri açıklama kısmında nasıl
# yazılmışsa burayada aynı şekilde yazıyoruz.
# tek score yazabiliyorum.
# verbose raporlama yapıyor. sayı arttıkça detayı artıyor. 0 ise yazmaz.
# 7*7*5 245 iterator yapar. 
# en iyi rmse score nu verecek parametrelerin kesişimini bana döndürüyor. 
# bütün parametrelerin birbirleri ile olan kesişiminin en iyi parametresini bize veriyor.
# ancak bunların advanced modellerde bize bir çalışma maliyeti olacak.
```

```python
grid_model.fit(X_train_scaled, y_train)
```

```python
grid_model.best_params_  # en iyi parametreleri döndürdü. gitti yine Lasso değerlerini yakaladı.
```

```python
pd.DataFrame(grid_model.cv_results_)  

# her iterasyonda en iyi rmse değerlerini ve bunların ortalamasını döndürdü. 
# indeksi 6 olan parametrenin en iyi score olduğunu buradan görebiliyorum.
# bu indeksin scorenu döndürdü.
```

```python
grid_model.best_index_ # hangi indekste en iyi score verdiğini gösteriyor.
```

```python
grid_model.best_score_
```
