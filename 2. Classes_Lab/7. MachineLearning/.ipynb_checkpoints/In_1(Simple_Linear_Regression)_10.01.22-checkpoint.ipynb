## Splitting the dataset into X(independent variables) and y (dependent variable)

```python
# y_pred = b1X + b0
```

```python
X= df["total_spend"]  # baÄŸÄ±msÄ±z deÄŸiÅŸken
y= df["sales"]   # baÄŸÄ±mlÄ± deÄŸiÅŸken
```

## Determination of coefficients (slope and intercept)

polyfit func. sadece burada iÅŸlevini gÃ¶stermek iÃ§in kullanÄ±yoruz. Ä°Ã§ine X ve y yi verince bana slope(ilk deÄŸer), intercept(ikinci deÄŸer) dÃ¶ndÃ¼rÃ¼r.

```python
np.polyfit(X, y, deg=1) 
```

```python
slope, intercept = np.polyfit(X, y, deg=1)
```

```python
print("slope    : {}".format(slope))
print("intercept:", intercept)
```

## Why do we use the least squares error method to find the regression line that best fits the data?

```python
# y_pred = b1X + b0

b1, b0 = np.polyfit(X, y, deg=1)
```

```python
print("b1 :", b1)
print("b0 :", b0)
```

```python
y_pred = b1*X + b0   # b2*X2 + b1*X + b0
```

```python
values = {"actual": y, "predicted": y_pred, "residual":y-y_pred, "LSE": (y-y_pred)**2}
df_2 = pd.DataFrame(values)
df_2
```

```python
df_2.residual
```

eÄŸer sadece residuallarÄ± toplarsam bu durumda sonuÃ§ gÃ¶rÃ¼ldÃ¼ÄŸÃ¼ gibi 0 olur ki bu durumda best fit line seÃ§miÅŸ olduÄŸumuz anlamÄ±na gelir.

LSE toplamÄ±na baktÄ±ÄŸÄ±mda ise Ã§izilen LSE line larÄ± arasÄ±nda bana en dÃ¼ÅŸÃ¼k score veren line bu olduÄŸunu biliyorum.

```python
df_2.residual.sum().round()
```

```python
df_2.LSE.sum()
```

## Prediction with simple linear regression

```python
potential_spend = np.linspace(0, 500, 100)
potential_spend
```

```python
predicted_sales_lin = b1* potential_spend + b0
predicted_sales_lin
```

```python
plt.plot(potential_spend, predicted_sales_lin)
sns.scatterplot(x='total_spend', y='sales', data=df)
```
