# Metin Görselleştirme

```python
import pandas as pd 
data = pd.read_csv("train.tsv",sep = "\t")

# train.tsv dosyasını kullanıyoruz.
```

```python
data = data[:5000]  # veri seti büyük olduğundan dolayı ilk 5000 satırı aldık.
data.head()
```

```python
data.info()
```

```python
#buyuk-kucuk donusumunu yaptık.

data['Phrase'] = data['Phrase'].apply(lambda x: " ".join(x.lower() for x in x.split()))

data['Phrase']
```

```python
#noktalama işaretlerini sildik.

data['Phrase'] = data['Phrase'].str.replace('[^\w\s]','')  

data['Phrase']
```

```python
#sayıların sildik.

data['Phrase'] = data['Phrase'].str.replace('\d','')

data['Phrase']
```

```python
#stopwords( Englsh) çıkardık.

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
sw = stopwords.words('english')
data['Phrase'] = data['Phrase'].apply(lambda x: " ".join(x for x in x.split() if x not in sw))

data['Phrase']
```

```python
#seyreklerin silinmesi

sil = pd.Series(' '.join(data['Phrase']).split()).value_counts()[-100:]
data['Phrase'] = data['Phrase'].apply(lambda x: " ".join(x for x in x.split() if x not in sil))

data['Phrase']
```

```python
#lemmitazition

from textblob import Word
#nltk.download('wordnet')
data['Phrase'] = data['Phrase'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()])) 
```

```python
data['Phrase']
```

```python
#buyuk-kucuk donusumu
df['text'] = df['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))

#noktalama işaretleri
df['text'] = df['text'].str.replace('[^\w\s]','')

#sayılar
df['text'] = df['text'].str.replace('\d','')

#stopwords
import nltk

#nltk.download('stopwords')
from nltk.corpus import stopwords
sw = stopwords.words('english')
df['text'] = df['text'].apply(lambda x: " ".join(x for x in x.split() if x not in sw))

#seyreklerin silinmesi
sil = pd.Series(' '.join(df['text']).split()).value_counts()[-1000:]
df['text'] = df['text'].apply(lambda x: " ".join(x for x in x.split() if x not in sil))

#lemmi
from textblob import Word

#nltk.download('wordnet')
df['text'] = df['text'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()])) 
```

## Değişken Mühendisliği

* Count Vectors
* TF-IDF Vectors (words, characters, n-grams)
* Word Embeddings : kelimenin yoğunluk haritası çıkarılıyor.

Tüm veri setindeki unique kelimeler çıkarılacak ve bunların herbirinin her bir satırda geçme sıklıkları yer alacak.

Herbir satır bir dokümanı, her bir sütunda kelimelerin burada geçme sıklıklarını ifade eder. Buna count vectors denir.

TF(t) = (Bir t teriminin bir dökümanda gözlenme frekansı) / (dökümandaki toplam terim sayısı) 

IDF(t) = log_e(Toplam döküman sayısı / içinde t terimi olan belge sayısı)
