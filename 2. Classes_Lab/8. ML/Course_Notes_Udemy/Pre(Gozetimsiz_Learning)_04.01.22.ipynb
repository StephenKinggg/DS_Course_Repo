# Hiyerarşik Kümeleme

```python
df = pd.read_csv("USArrests.csv").copy()
df.index = df.iloc[:,0]
df = df.iloc[:,1:5]
#del df.index.name
df.head()
```

```python
df = df.rename_axis(None, axis=0)
```

```python
from scipy.cluster.hierarchy import linkage
hc_complete = linkage(df, "complete")
hc_average = linkage(df, "average")
hc_single = linkage(df, "single")
```

```python
dir(hc_complete)
```

```python
from scipy.cluster.hierarchy import dendrogram

plt.figure(figsize=(15, 10))
plt.title('Hiyerarşik Kümeleme - Dendogram')
plt.xlabel('Indexler')
plt.ylabel('Uzaklık')
dendrogram(
    hc_complete,
    leaf_font_size=10
);  #  ; koymazsak değerler gelir.
```

```python
from scipy.cluster.hierarchy import dendrogram

plt.figure(figsize=(15, 10))
plt.title('Hiyerarşik Kümeleme - Dendogram')
plt.xlabel('Indexler')
plt.ylabel('Uzaklık')
dendrogram(
    hc_complete,
    truncate_mode = "lastp",
    p = 4,  # buradaki sayı ile oynayarak kaç kümeye ayrılacağını görebiliriz.
    show_contracted = True
);
```

## Optimum Küme Sayısı

# PCA

```python
df = pd.read_csv("USArrests.csv").copy()
df.index = df.iloc[:,0]
df = df.iloc[:,1:5]
#del df.index.name
df.head()
```

```python
df = df.rename_axis(None, axis=0)
```

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
from sklearn.preprocessing import StandardScaler

df = StandardScaler().fit_transform(df) # standartlaştırma işlemi yapmış olduk.
df[0:5,0:5]
```

```python
# yukarıdaki df aşağıdaki gibi 3 değişken ile temsil ettik.

from sklearn.decomposition import PCA
pca = PCA(n_components = 3)   # bileşen sayısını giriyoruz. 2 değişken ihtiyacı karşıladıgından dolayı yeterli olduğundan diğerlerine gerek yok.
pca_fit = pca.fit_transform(df)
```

```python
bilesen_df = pd.DataFrame(data = pca_fit, 
                          columns = ["birinci_bilesen","ikinci_bilesen","ucuncu_bilesen"])
```

```python
bilesen_df.head()
```

```python
pca.explained_variance_ratio_ # üçüncü bileşenin açıklanan varyansı açıklama oranı düşük olduğundan sadece bu iki bileşen almamız yeterli.
```

```python
pca = PCA().fit(df)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_)) # bu grafik bize bileşen sayısına karar vermemizde bize fayda sağlıyor.
```
