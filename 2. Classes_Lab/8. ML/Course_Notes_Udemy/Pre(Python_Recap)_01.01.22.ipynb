```python
#Veri Yapıları - Tuple : Kapsayıcıdır(farklı veri tipleri içerir), sıralıdır(indeks işlemleri yapılabilir)
# ancak liste lerden farkı değiştirilemez oluşlarıdır.

# Mesela bazı durumlarda veri yapıları sabit dursun ve ben değişiklik olup olmamasından endişe etmeyeyim, 
# bir referans listesi oluşturma gibi bir durum söz konusu olduğunda kullanırız.

#Tuple Olusturma : aşağıdaki gibi oluşturabiliriz.

t = ("ali","veli",1,2,3.2,[1,2,3,4])

t
```
