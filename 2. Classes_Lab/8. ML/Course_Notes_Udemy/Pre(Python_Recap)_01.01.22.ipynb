```python
#Setler - Klasik Kume Islemleri

# =============================================================================
# difference() ile iki kumenin farkini ya da "-" ifadesi
# intersection() iki kume kesisimi ya da "&" ifadesi
# union() iki kumenin birlesimi
# symmetric_difference() ikisinde de olmayanlari.
# =============================================================================
```

```python
#difference

set1 = set([1,3,5])
set2 = set([1,2,3])

print(set1.difference(set2))

print(set2.difference(set1))

print(set1.symmetric_difference(set2))  # ikisinde de olmayanları getirir.
```

```python
print(set1 - set2)
print(set2 - set1)
```

```python
#intersection & union

set1 = set([1,3,5])
set2 = set([1,2,3])

print(set1.intersection(set2))
print(set2.intersection(set1))
```

```python
kesisim = set1 & set2  # sonuçları saklamak istersek bunları yeni bir değişkene atıyoruz.
kesisim
```

```python
birlesim = set1.union(set2)
birlesim
```

```python
set1.intersection_update(set2)  # kesişimde gelen elemanları set1 update yaptık.
set1
```

```python
# mini uygulama

maaslar = [1000,2000,3000,4000,5000]

def maas_ust(x):   # üst maaşlar için zamlı maaş hesabı func yazdık.
    print(x*10/100 + x)

def maas_alt(x):   # alt maaşlar için zamlı maaş hesabı func yazdık.
    print(x*20/100 + x)

for i in maaslar:
    if i >= 3000:
        maas_ust(i)
    else:
        maas_alt(i)
```

```python
#break & continue : belirli bir şartı sağlayan ifadeler yakalandığında döngü bitirilmek veya bu durumu görmezden gelmek gerekebilir.
        
# burada maaşı 3000 e kadar olanlar ile ilgilendiğimizi düşünelim.
maaslar = [8000,5000,2000,1000,3000, 7000, 1000]

# dir(maaslar)  # maaşlara uygulanabilecek methodları görmek istedik.

maaslar.sort()
print(maaslar)

for i in maaslar:
    if i == 3000:
        print("kesildi")  # döngünün kesildiğini takip edebilmek için yazdık.
        break
    print(i)
```

```python
for i in maaslar:
    if i == 3000:
        continue  # bir değer kontrol ifadesine takıldığında bu değeri atlamak için kullanılır. Görüldüğü gibi 3000 gelmedi.
    print(i)
```

```python
# while: bu şort sağlandığı sürece döner.

sayi = 1

while sayi < 9:  # sayı 9 dan küçük olduğu sürece her iterasyonda buradaki sayıya 1 ilave ederek sayıya yeni bir atama yapıyor.
    sayi += 1
    print(sayi)
```

```python
# Sinif Nedir?
# İşlemlerimizi kolaylaştırmak, biraz daha basit işlemler yapabilmek adına kullanılır.

# Sinif Ozellikleri (Class attributes)
# Benzer özellikler, ortak amaçlar taşıyan, içerisinde method ve değişkenler bulunan yapılardır.

class VeriBilimci():  # class tanımlıyoruz.
    bolum = ''
    sql = 'Evet'
    deneyim_yili = 0
    bildigi_diller = []
```

```python
# Ornek ozellikleri

# Yukarıdaki hatayı gidermek için yani her bir instance ayrı özelliklerden oluştuğunu python vermemiz gerekir.

class VeriBilimci():  # hem class özelliklerini tanımladık. Hem de aşağıda instance özelliklerini tanımladık.
    bildigi_diller = ["R","PYTHON"]
    bolum = ''
    sql = ''
    deneyim_yili = 0      # self burada örneklemleri temsil eder.
    def __init__(self):  # her bir instance kendi içinde ayrı özelliklere sahip olduğunu sağlamak için
        self.bildigi_diller = []  # instanceler için tanımlama yapıyoruz.
        self.bolum = ''
```

```python
# Ornek Metodlari

# örneklem üzerinde çalışan yani her bir veribilimcinin öğrendiği yeni dili kaydetsin.

class VeriBilimci(): # 
    
    calisanlar = []  # çalışan isminde bir nesne yani object tanımladım.
    def __init__(self):  # örneklemlerin özelliklerini tanımladık.
        self.bildigi_diller = []
        self.bolum = ''
        
    def dil_ekle(self, yeni_dil) : # dil ekleme işini tek tek yapmak istemediğimizden dolayı.
        self.bildigi_diller.append(yeni_dil)  # örneklem özelinde ve class altında çalışıyor.
```

```python
# Miras Yapilari (inheritance)

# daha önce tanımladığımız bir class ın özelliklerini yeni class ta kullanmak istiyorsak bunu kullanırız.

class Employees():
    def __init__(self):
        self.FirstName = ""
        self.LastName = ""
        self.Address = ""


class DataScience(Employees):  # burada DataScience() yerine bunların örnek özelliklerini kullanmak için bunun için o sınıfın ismini yazıyoruz. 
    def __init__(self):
        self.Programming = ""

# veribilimci1 = DataScience()
# veribilimci1. dediğimizde hem kendi classının hem de employees sınıfının özellikleri gelir.

class Marketing(Employees):
    def __init__(self):
        self.StoryTelling = ""
```

```python
class Employee_yeni():  # yukarıdaki işi daha görsel bir şekilde yapmak istersek.
    def __init__(self, FirstName, LastName, Address):
        self.FirstName = FirstName
        self.LastName = LastName
        self.Address = Address


ali = Employee_yeni("a","b","c")
print(ali.FirstName)
print(ali.LastName)
print(ali.Address)
```
