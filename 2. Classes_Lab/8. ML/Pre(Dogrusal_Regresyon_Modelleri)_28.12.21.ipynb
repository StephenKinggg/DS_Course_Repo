```python
from sklearn import model_selection
```

```python
cv_10 = model_selection.KFold(n_splits = 10,  # cv: cross validation
                             shuffle = True,  # gruplara ayrılmadan önce verilerin karıştırılıp karıştırılmayacağını gösteriyor.
                             random_state = 1)
```

```python
lm = LinearRegression()
```

# Ridge Regresyon

## Model

```python
hit = pd.read_csv("Hitters.csv")
df = hit.copy()
df = df.dropna()
ms = pd.get_dummies(df[['League', 'Division', 'NewLeague']])
y = df["Salary"]
X_ = df.drop(['Salary', 'League', 'Division', 'NewLeague'], axis=1).astype('float64')
X = pd.concat([X_, dms[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.25, 
                                                    random_state=42)
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha = 0.1).fit(X_train, y_train)
```

```python
ridge_model
```

```python
ridge_model.coef_
```

```python
10**np.linspace(10,-2,100)*0.5   # 10 ile -2 arasında 100 tane değer getirdi.
```

```python
lambdalar = 10**np.linspace(10,-2,100)*0.5 

ridge_model = Ridge()  # object oluşturduk.
katsayilar = []        # array oluşturduk.

for i in lambdalar:
    ridge_model.set_params(alpha = i)    # her iterasyonda alfa değerlerini değiştiriyoruz. 
    ridge_model.fit(X_train, y_train)    # modeli fit ettik.
    katsayilar.append(ridge_model.coef_) # her iterasyonda elde edilen katsayılar ile doldurduk.
  # her iterasyonda 19 tane değer oluşacak.  

    
ax = plt.gca()
ax.plot(lambdalar, katsayilar) 
ax.set_xscale('log') 

plt.xlabel('Lambda(Alpha) Değerleri')
plt.ylabel('Katsayılar/Ağırlıklar')
plt.title('Düzenlileştirmenin Bir Fonksiyonu Olarak Ridge Katsayıları');
```

## Tahmin

```python
y_pred = ridge_model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))  # test hatasını bulduk.
```

```python
lambdalar[0:5]
```

```python
from sklearn.linear_model import RidgeCV
ridge_cv = RidgeCV(alphas = lambdalar, 
                   scoring = "neg_mean_squared_error",
                   normalize = True)
```

```python
ridge_cv.fit(X_train, y_train)
```

```python
ridge_cv.alpha_
```

```python
ridge_tuned = Ridge(alpha = ridge_cv.alpha_, 
                   normalize = True).fit(X_train,y_train)
```

```python
np.sqrt(mean_squared_error(y_test, ridge_tuned.predict(X_test)))
```

# Lasso Regresyon

## Model

```python
hit = pd.read_csv("Hitters.csv")
df = hit.copy()
df = df.dropna()
ms = pd.get_dummies(df[['League', 'Division', 'NewLeague']])
y = df["Salary"]
X_ = df.drop(['Salary', 'League', 'Division', 'NewLeague'], axis=1).astype('float64')
X = pd.concat([X_, dms[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.25, 
                                                    random_state=42)
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso_model = Lasso(alpha = 0.1).fit(X_train, y_train)
```

```python
lasso_model
```

```python
lasso_model.coef_
```

```python
lasso = Lasso()
lambdalar = 10**np.linspace(10,-2,100)*0.5 
katsayilar = []

for i in lambdalar:
    lasso.set_params(alpha=i)  # lasso model nesnesinin parametresini bu alpha değeri için fit ederek bir parametre atar.
    lasso.fit(X_train, y_train)  # modeli fit eder.
    katsayilar.append(lasso.coef_)  # her bir lambda için değişken kadar katsayı hesaplayarak bunları ekliyor.
    
ax = plt.gca()
ax.plot(lambdalar*2, katsayilar)
ax.set_xscale('log')
plt.axis('tight')
plt.xlabel('alpha')
plt.ylabel('weights')
```

## Tahmin
