!!! ÖNEMLİ:

28*28 boyutunu B/s anlatmamız gerekir. Bunun için bunu 3 boyutlu hale getirdik. 3 boyut olayını tüm X degerleri için yaptık ancak bunu y değerleri için yapmıyoruz çünkü y değerleri scalar değerlerdir. Bunun için keras.utils kullanarak y_train kaç tane class olduğunu biliyorsak onu da belirterek kategorige çeviriyoruz aşağıdaki gibi. Burada bir anlamda onehot encoding yapmış olduk.

Bu sekilde model artik regression degil, classification yapmasi gerektigini bilecek.

```python
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1) 
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1)
```

```python
X_train.shape, X_test.shape
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
Y_train = to_categorical(y_train, 10)  
# kaç class olduğunu biliyorsak bunu yazmamızın bizden sonra bakanlar için faydalı olacağını unutmayalım.
Y_test = to_categorical(y_test, 10)
Y_test.shape
```

Kerasın sequential dışında functional api da var. Functionalda birbirinden inherite ederken sequential da ise layerları arka arkaya diziyoruz. Sequential kullanımı daha kolay.

İlk olarak sequential objesini tanımladık. Sonra sırasıyla bunun içine layer eklemeye başlıyoruz. Icine ilk olarak Convolution layer' i ekledik. Bunun icine de __input_shape__ ekledik. Bu, arka planda input_layer' in gorevini de yapar. Input_shape' i bastan belirledigimiz icin model.summary ile modelin ozetine bakabiliriz. Input boyutu 28x28 idi biz bunu 3 boyutlu hale getirerek 28x28x1 yaptik. Input shape burada belirlediğimiz için modeli fit etmeden önce summary bakabiliriz.

32 tane filtre(ANN' deki nöron sayisi ile esdeger) belirledik. __kernel_size__ ile her bir filtrenin boyutunu 3x3 olarak belirledik. (Padding ve stride parametrelerini default olarak kullandigimiz icin belirtmedik)

Default olarak padding valid ve stride 1 olarak kullanıyoruz.

RNN farklı activation func kullanılabilirken CNN de Activation func. olarak her zaman ReLu kullanıyoruz.

Aşağıda sırasıyla Convolution layer , maxpooling, flatten, dense layer yani fully connected ve sonra output layer kullanıyoruz. 

Convolution' dan sonra maxpooling ve sonrasında ise flatten ekledik. __Flatten,__ pool layer' dan aldigi tüm degerleri tek row haline getirdik. Bundan sonrası ANN ile aynı.

Bu islemden sonra __Dense__ kismini tanimladik. 128 nöronlu, activation' i ReLu olan bir dense layer kullandik (fully connected asamasi). 

Sonrasinda yine dense ile 10 nöronlu output layer' i tanimladik ve activation' i softmax sectik.

Son olarak loss function' i categorical_crossentropy, optimizer' i adam olarak belirledik ve metrics olarak da accuracy' i takip edecegimizi soylemis olduk. 

Dense output ta kullanacağız.

En büyük başarıyı veren optimizer "adam" dır. Optimizer demek aslında bizim gradient descent uygulayan algoritmadır. Adam' in en buyuk ozelligi; arkada gradient descent algoritmasini uygularken önceki batch'te aldigi türevlerin agirlikli ortalamasini alarak devam etmesidir. Yokuş aşağı inerken sapma yapmasını engeller. Her türev alma isleminden sonra min noktaya yaklastigini dusunurse learning_rate' i otomatik olarak düşürür ki min. noktayi atlamasin.

Türev eğim ile alakalı olduğundan eğim düştükçe min. yaklaştığımızı anlamamız gerekir.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten
```

```python
model = Sequential()

model.add(Conv2D(filters=32, kernel_size=(3,3), input_shape=(28, 28, 1), activation='relu'))

model.add(MaxPool2D(pool_size=(2, 2)))

model.add(Flatten())

model.add(Dense(128, activation='relu'))

model.add(Dense(10, activation='softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

Dense layer da çok fazla parametre olduğunu görüyoruz eğitmek için. 

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=2)
```

!!! Çok Önemli:

2 epochs üst üste val_loss değeri cıvıtırsa durduruyor.

Batch size genellikle 8-64 arası verilir. Bunu yüksek tutmak memory çok zorlar.

Early stop kullanacaksak epoch bir önemi yok. Early stop kullanmayacaksak modeli çok iyi regülüze etmemiz gerekir. 

Early stop skoru düşürmez, sadece overfittingi engeller. Bazen sadece patiance değeri düşük verilirse, o zaman belki daha düşük bir min. değere gidecekken gitmeyebilir.

Seed kullanılmadığında herkesin epoch sayısı değişebilir. 

Küçük data setlerinde her seferinde random olarak başka bir yerden başlaması güzel ancak büyük data setlerinde bu tercih edilen bir durum değildir. Çünkü eğitim uzun zaman alıyor. Bunu her seferinde tekrar başlatmak iyi olmayabilir ve aynı zamanda mümkün de olmayabilir.

```python
model.fit(X_train, Y_train, batch_size=32, epochs=10, validation_data=(X_test, Y_test), callbacks=[early_stop])
```

GPU lar mimarisi gereği seed koysaydak parelel işlemi yaptığımızdan dolayı aynı sonucu almak imkansız gibidir.

```python
plt.figure(figsize=(10,6))
plt.plot(summary.accuracy)
plt.plot(summary.val_accuracy)
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.legend('labels');
```

Modelimizi test setleri üzerinde evaluate ettik; loss ve acccuracy degerlerini aldik. Ancak son kısımda validation data olarak test setini kullandığımızdan dolayı orada çıkan loss değeri ile aşağıda çikan loss değerinin aynı olması gerekir.
