Experimental data used to create regression models of appliances energy use in a low energy building.
Data Set Information:

The data set is at 10 min for about 4.5 months. The house temperature and humidity conditions were monitored with a ZigBee wireless sensor network. ### Define a Function for Creating Time Steps 

daha öncesinde time series generator ile şu dataları al ve tahmin yapma işlemini uygulamıştık. 

Şimdi time series generatorun yaptığı işi manuel olarak yapmak için aşağıdaki func tanımladık. 

Burada şunu yapacağım: mesela 3 satırın tüm verisini alarak bana target feature tahmini yapacak. sonra kaydırarak bir sonraki 3 satırı alacağız. bu şekilde devam edecek. burada 

İki yöntem kullanacağız öncesinde tahmin sonucu oluşan target feature değeri ve buna denk gelen diğer feature değerlerini kullanma ve direk yeni oluşan target feature değeri yerine reel feature değerini kullanma.

ML ve Diger DL algoritmalarında buradan farklı olarak ilk önce target ve diğer featureları ayırıp bunlardan mesela ilk 3 satırı kullanarak target feature tahmin ediyordum.

Buradaki yani RNN deki fark ise tahmin yaparken target feature değerini de kullanıyor.

Bu işlemi ise aşağıdaki gibi;

for döngüsü X uzunluğundan time steps çıkarıyorum. çünkü length ne kadar ise son kısımda bu kadara işlem yapamıyordu. X in yani train datasetinin i den i+time steps kadar olan kısmın valuelarını alıyorum ve bunu birdeğişkene atıyorum. bu değişkenleri kaydırarark 24 lü paketler halinde  tanımladığım boş listeye atıyor. burada mesela i=0 iken v değişkenin ilk 0-23 arasındaki 24 satır, y değişkenine ise targetin 24 satırı geliyor. bunu datanın sonuna kadar uyguladık. bu işlem sonucunu ise array dönüştürüyoruz.

```python
def create_dataset(X, y, time_steps=1):
    """ Create data sequence
    
    Arguments:
        * X: time-series data
        * y: target value
        * time_steps: Used to create input sequence of timesteps
    
    Returns:
        * input_sequence: Numpy array of sequences of time-series data
        * output: Numpy array of output i.e. next value for respective sequence
    
    """
    Xs, ys = [], []
    for i in range(len(X) - time_steps):  # burada xtrainden 0-23 indeksleri
        v = X.iloc[i:(i + time_steps)].values
        Xs.append(v)        
        ys.append(y.iloc[i + time_steps])  # burada xtest 24 indeksi
    return np.array(Xs), np.array(ys)   # bana bir array döndürdü.
```

Aşağıdaki gibi train ve test datalarımı bölüyorum.

```python
time_steps = 24   # yani length 24.

# reshape to [samples, time_steps, n_features]

X_train, y_train = create_dataset(train, train.Appliances, time_steps)
X_test, y_test = create_dataset(test, test.Appliances, time_steps)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

## Model Building with LSTM

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Bidirectional, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
n_features = X_train.shape[2]
```

Görüldüğü gibi model benden 24*10 luk bir veri bekliyor.

Bidirectional, buradaki 24 lük serilerimi önce 1-24 e kadar sonrada 24-1 kadar alıp bunların sonuçlarının toplamını bir sonraki katmana aktarma işlemini yapıyor. buradaki ağırlıkların toplamını döndürerek bir sonraki layer gönderir.

output layer da 1 tane nöron olacak. 

early stop patience=5 deyince en sondan 5. değer benim en iyi değerim. daha sonra daha iyiye gitmeyip durmuş model. en iyi ağırlıkların olduğu yeri al demesen muhtemelen en sondakini alacak ve buda muhtemelen en iyiye göre biraz daha düşük olacaktır. bundan dolayı restore_best_weights = True early stop içine eklemekde fayda var.

!!! önemli:

yukarıda loss değeri epoch içindeki batchlerin ortalaması oluyor. ancak val_loss ise ilk epoch sonunda oluştuğundan dolayı ilk olarak yüksek başlar ancak zamanla loss artarken val_loss azalır. 

batch size çok büyük almamak gerekir. 

time series generator shuffle=True(default değer budur zaten) demek her epochta bir benim datamı karıyor ve ona sıraya göre datamı modele sokuyor. benim datam burada 24 lü paketler halinde olduğundan dolayı bu 24 lük paketlerimin yerinin değişmesinin bir anlamı yok. Paket içinde bir değişiklik olmuyor. ancak paketlemeyi yapana kadar datanın sıralaması önemli. Timeseries generator ile yaparken bu paketleme işini kendisi yapıyordu. daha sonra fit dediğimizde de shuffle=True bırakmıştık. Sonuç olarak paketleyene kadar shuffle=False paketledikten sonra shuff=True kullanmalıyız. 

```python
model = Sequential()

model.add(Bidirectional(LSTM(units=100, input_shape=(time_steps, n_features))))

#model.add(Dropout(rate=0.2))

model.add(Dense(units=1))

model.compile(optimizer='adam', loss='mse')
```

```python
early_stop = EarlyStopping(monitor = 'val_loss', patience = 5, restore_best_weights = True)
```

```python
model.fit(X_train, y_train, 
          epochs=30, 
          batch_size=24, 
          validation_split=0.3,
#          shuffle=False,
          callbacks=[early_stop])
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

## Evaluate on Test Data

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def eval_metrics(actual, pred):
    rmse = np.sqrt(mean_squared_error(actual, pred))
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    score = r2_score(actual, pred)
    return print("r2_score:", score, "\nmae:", mae, "\nmse:",mse, "\nrmse:",rmse)
```

```python
forecast = []
# Replace periods with whatever forecast length you want
periods = 24

first_eval_batch = full.values[-time_steps:]
current_batch = first_eval_batch.reshape((1, time_steps, full.shape[1]))
features = df_fake.values.reshape((df_fake.shape[0], 1, df_fake.shape[1]))

for i in range(periods):
    
    # get prediction 1 time stamp ahead 
    current_pred = model.predict(current_batch)[0][0]
    
    # store prediction
    forecast.append(current_pred) 
    new = features[i][0].tolist()
    new.insert(0,current_pred)   # burada predict sonucu olan değeri verdik.
    new = np.array(new).reshape(1,1,full.shape[1])
    
    # update batch to now include prediction and drop first value
    current_batch = np.append(current_batch[:, 1:, :], new, axis = 1)
```
