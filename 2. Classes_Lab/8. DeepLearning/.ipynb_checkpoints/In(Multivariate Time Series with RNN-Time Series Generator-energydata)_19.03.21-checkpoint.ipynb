# Dataset Info

Appliance Energy Usage: A Multivariate Time Series Forecasting Example

Experimental data used to create regression models of appliances energy use in a low energy building.
Data Set Information:

The data set is at 10 min for about 4.5 months. The house temperature and humidity conditions were monitored with a ZigBee wireless sensor network. Each wireless node transmitted the temperature and humidity conditions around 3.3 min. Then, the wireless data was averaged for 10 minutes periods. The energy data was logged every 10 minutes with m-bus energy meters. Weather from the nearest airport weather station (Chievres Airport, Belgium) was downloaded from a public data set from Reliable Prognosis (rp5.ru), and merged together with the experimental data sets using the date and time column. Two random variables have been included in the data set for testing the regression models and to filter out non predictive attributes (parameters).
Original source of the dataset:

http://archive.ics.uci.edu/ml/datasets/Appliances+energy+prediction

# Data Pre-Processing

## Feature Engineering
- Let's reduce the number of feature
- Let's resample the dataset by hours

```python
df.columns
```

Sıcaklık değerlerinin ortalamasını alıyorum. Aynı şeyi nem değerleri içinde yapıyorum. Amaç feature sayısını azaltmak.

```python
df=df.rename(columns = {'T6':'T_outside', 'RH_6':'RH_outside'})
df["T_inside"] = (df["T1"] + df["T2"] + df["T3"] + df["T4"] + df["T5"] + df["T7"] + df["T8"] + df["T9"])/8
df["RH_inside"] = (df["RH_1"] + df["RH_2"] + df["RH_3"] + df["RH_4"] + df["RH_5"] + df["RH_7"] + df["RH_8"] + df["RH_9"])/8
```

```python
drop_list = ['T1', 'RH_1', 'T2', 'RH_2', 'T3', 'RH_3', 'T4', 'RH_4', 'T5', 'RH_5',
             'T7', 'RH_7', 'T8', 'RH_8', 'T9', 'RH_9', 'T_out', 'RH_out', 'rv1', 'rv2']
df.drop(drop_list, axis = 1, inplace = True)
```

```python
df.info()
```

```python
df_a = df[["Appliances", "lights"]]
df_b = df.drop(["Appliances", "lights"], axis = 1)
```

Datayı 10 dk lık bıraksaydık sequencelerin uzunluğu 144 olacaktı. LSTM in aklında tutması için uzun bir periyot olacaktı. LSTM datanın önemli kısmını aklında tutuyordu. Hangi veri daha önemli hangisini unutması gerektiğini eğitilerek anlıyordu. Burada data 10 dklık olduğundan dolayı cycle günlük almak daha mantıklı olabilir. 

Datetime datalar ile çalışınca resample yapabiliyoruz. Bizden beklenen işe göre buna karar vereceğiz ancak burada amaç eğitim olduğundan dolayı datamı time series datalarda kullanabildiğim resample kullanıyorum. 

Datayı saatliğe çevirmek için H kısaltmasını kullandım. Sonra Appliances ile lights değişkenlerinin toplamını alırken dışındakilerin ise ortalamasını alıyorum. indeks sütunundaki datayı H ya göre ayarlıyor. Bu durumda 144 satırı saatlik alınca yani 6 tane 10 dk.lık bölünce 24 satıra düşürdük.

```python
df_a = df_a.resample("H").sum()
df_b = df_b.resample("H").mean()
```

```python
df = pd.concat([df_a, df_b], axis =1)
df.head()
```

```python
df.shape
```

Let's also round off the data, to one decimal point precision, otherwise this may cause issues with our network (we will also normalize the data anyways, so this level of precision isn't useful to us)

```python
df = df.round(2)
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(df.corr(), annot=True);
```

```python
df['Appliances'].plot(figsize=(18,6))
plt.xticks(rotation=90);
```

Appliance yoğunlaşmak istiyorum. Ancak yukarıdan net bir şey göremediğim için bunu 3 güne düşürdüm.

```python
df['Appliances'].loc["2016-05-01 00:00:00" : "2016-05-03 23:00:00"].plot(figsize=(18,6))
plt.xticks(rotation=90);
```

### Train Test Split

```python
len(df)
```

```python
# How many rows per day?
24
```

```python
test_days = 7
```

```python
test_ind = test_days*24
```

```python
test_ind
```

```python
train = df.iloc[:-test_ind]
test = df.iloc[-test_ind:]
```

```python
train.shape, test.shape
```

### Scale Data

```python
from sklearn.preprocessing import MinMaxScaler, RobustScaler
```

```python
scaler = MinMaxScaler()
```

```python
train_scaled = scaler.fit_transform(train)
test_scaled = scaler.transform(test)
```

## Create the Model

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Bidirectional, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
length = 24 # Length of the output sequences (in number of timesteps)
batch_size = 1 #Number of timeseries samples in each batch
generator = TimeseriesGenerator(train_scaled, train_scaled, length=length, batch_size=batch_size)
validation_generator = TimeseriesGenerator(test_scaled, test_scaled, length=length, batch_size=batch_size)
```

```python
len(train_scaled), len(generator), len(test_scaled), len(validation_generator)
```

```python
train_scaled.shape
```

```python
n_features = train_scaled.shape[1]
```

Time series larda generator sadece appliance üzerine çalışıp da diğer featureların appliance üzerine etkisini görmezden gelen bir model oluşturmuyor. her feature kendi içinde değerlendirip onun için bir prediction yapıyor. yani bunu her feature için yapıyor. kaç tane feature varsa her feature için geriye doğru tahmin yapıyor. 

Timeseries generator yani böyle multivariate datalar için uygun değil.

Daha önce reg.probleminde outputa birden fazla nöron sadece bir tane koymuştuk. burada ise birden fazla değişken yani nöron söz konusu. hepsi için bir reg. yapıyor.

```python
model = Sequential()

model.add(LSTM(100, input_shape=(length, n_features)))

model.add(Dense(n_features))  # her bir feature için reg. yapacağından dolayı.

model.compile(optimizer='adam', loss='mse')
```

```python
early_stop = EarlyStopping(monitor='val_loss',patience=3)
```

```python
model.fit_generator(generator,
                    epochs=10,
                    validation_data=validation_generator,
                    callbacks=[early_stop])
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

## Evaluate on Test Data

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def eval_metrics(actual, pred):
    rmse = np.sqrt(mean_squared_error(actual, pred))
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    score = r2_score(actual, pred)
    return print("r2_score:", score, "\nmae:", mae, "\nmse:",mse, "\nrmse:",rmse)
```

```python
train_scaled
```

```python
predictions_scaled = []

first_eval_batch = train_scaled[-length:]
current_batch = first_eval_batch.reshape((1, length, n_features))

for i in range(length):
    
    # get prediction 1 time stamp ahead
    current_pred = model.predict(current_batch)
    
    # store prediction
    predictions_scaled.append(current_pred[0]) 
    
    # update batch to now include prediction and drop first value
    current_batch = np.append(current_batch[:, 1:, :], [current_pred], axis=1)
```

```python
predictions_scaled
```

```python
eval_metrics(test_scaled[:length], predictions_scaled)

# bu arraylarin ortalaması gibi düşünelim. bunları tek tek karşılaştırmak istiyorum. 
```

Multivariate datalarda time serie generator kullanmak çok mantıklı.

## Inverse Transformations and Compare

!!! önemli:

Aşağıda predict yaparken mesela length kadar alınan tüm satır ve sütunları kullanarak diğer featureları için predict yapıyor. Bunu yaparken tüm featureları kullanıyor. Yani tüm featurelar birer target sütun gibi düşünüp bunlar için ayrı predict yapıyor ancak bu durumda kendi içinde bir birim hata yaparken bu tüm featurelarda sonraki predictlerde daha da artıyor.

time series generatorleri multivariate datalarda yani bir bağımlı birkaç tane de bağımsız değişken varsa bunlarda kullanmak çok mantıklı olmayabilir. bu durumda manuel yapmak gerekir.

ancak şöyle bir durumda yani 2 feature var bunlar birbirine bağımlı. bir ürünün fiyat değişimi ile arz talep değişimi gibi. bu durumlarda bunların birindeki değişim diğerini etkiliyor ise bu durumlurda time series generatorlar kullanmak daha uygun olabilir.

```python
predictions = scaler.inverse_transform(predictions_scaled)
predictions = pd.DataFrame(data=predictions, columns=test.columns)
predictions
```

Her feature forecastleri aldık. Bunları test datamın ilk başındaki 24 data ile karşılaştırıyoruz. aşağıdaki for döngüsü ile test seti ile prediction her bir değerini alıp eval metrics ile karşılaştırıyorum.

```python
test_new = test[:length]
```

```python
for i in test_new.columns:
    print(i)
    eval_metrics(test_new[i], predictions[i])
    print("--"*15)
```

```python
index = 0
plt.figure(figsize = (20, 30))

for i in test.columns:
    index += 1
    test_eva = pd.DataFrame(test_new[i])
    test_eva["predictions"] = predictions[i].values
    plt.subplot(5, 2, index)
    sns.lineplot(x = test_eva.index, y = test_eva[i])
    sns.lineplot(x = test_eva.index, y = test_eva.predictions)
    plt.xticks(rotation=90);
```

bizim data setimizde time olması demek onun zamana göre bir tahmin yapabileceğini göstermez. featurelar ile time arasında bir ilişki olmalı. zamana göre bir seasonality olması gerekir. mesela burada appliance sadece zamanla değişen bir feature değil. bunlara etki eden başka hususlarda var ancak model bunları yansıtamamış. 

burada tüm featureları kendi içinde ve time göre değerlendirdiğinden dolayı bir seasonality yakalayamadı.

aşağıda olduğu gibi sadece appliance feature alarak aynı işlemleri yaptım ve sonuca baktım.

aslında her feature için bir tahmin yapması aslında bizim hem istemediğimiz bir şey hemde gerek yok. aslında sadece ortalama tahmini enerji tüketimi hedefimiz.
