### Proje 1 

Kodlama Egzersizimizde yazdığımız Kumanda Sınıfına ek olarak metodlar ve özellikler eklemeye çalışın.

```python
import random
import time

class RemoteControl():
    
    def __init__(self, TV_durum="Kapalı", TV_ses=0, Kanal_List=["TRT"], Kanal="Trt"):
        self.TV_durum = TV_durum
        self.TV_ses = TV_ses
        self.Kanal_List = Kanal_List
        self.Kanal = Kanal
    
    def TV_ac(self):
        
        if (self.TV_durum=="Açık"):
            print("TV zaten açık durumda.")
        else:
            print("TV açılıyor...")
            self.TV_durum = "Açık"    
        
    def TV_kapat(self):   
        
        if (self.TV_durum=="Kapalı"):
            print("TV zaten kapalı durumda.")
            
        else:
            print("TV kapanıyor...")
            self.TV_durum = "Kapalı"
    
    def ses_ayarları(self):
        
        while True:
            cevap = input("Sesi azalt: '<'\nSesi arttır: '>'\nÇıkış : çıkış")
            
            if(cevap=="<"):
                if (self.TV_ses !=0):
                    
                    self.TV_ses -=1
                    print("ses:", self.TV_ses)
                    
            elif (cevap==">"):
                if (self.TV_ses!=31):
                    
                    self.TV_ses +=1
                    print("ses", self.TV_ses)
            else:
                print("Ses güncellendi:",self.TV_ses)
                break
    
    def kanal_ekle(self, kanal_ismi):   
                 
        print("Kanal ekleniyor...")
        time.sleep(1)
        
        self.Kanal_List.append(kanal_ismi)
        print("Kanal eklendi.")
    
    def rastgele_kanal(self):   
         
        rastgele = random.randint(0,len(self.Kanal_List)-1)
        
        self.Kanal = self.Kanal_List[rastgele]
        print("Şu anki kanal:", self.Kanal)
                    
    def __len__(self):
        
        return len(self.Kanal_List)
    
    def __str__(self):
    
        return "TV Durumu: {}\nTV Ses: {}\nKanal Listesi: {}\nŞuanki Kanal: {}\n".format(self.TV_durum, self.TV_ses, self.Kanal_List, self.Kanal)
                    

kumanda = RemoteControl()
                    
"""
Televizyon Uygulaması


1. Tv Aç

2. Tv Kapat

3. Ses Ayarları

4. Kanal Ekle

5. Kanal Sayısını Öğrenme

6. Rastgele Kanala Geçme

7. Televizyon Bilgileri

Çıkmak için 'q' ya basın.
"""  
while True:

    işlem = input("İşlemi Seçiniz:")

    if (işlem == "q"):
        print("Program Sonlandırılıyor...")
        break

    elif (işlem == "1"):
        kumanda.TV_ac()
                    
    elif (işlem == "2"):
        kumanda.TV_kapat()

    elif (işlem == "3"):
        kumanda.ses_ayarları()

    elif (işlem == "4"):
        kanal_isimleri = input("Kanal isimlerini ',' ile ayırarak girin:")

        kanal_listesi = kanal_isimleri.split(",")

        for eklenecekler in kanal_listesi:
            kumanda.kanal_ekle(eklenecekler)
   
    elif (işlem == "5"):

        print("Kanal Sayısı:",len(kumanda))

    elif (işlem == "6"):
        kumanda.rastgele_kanal()
   
    elif (işlem == "7"):
        print(kumanda)

    else:
        print("Geçersiz İşlem......")   
```

### Proje 2 

Bir tane "Bilgisayar" sınıfı oluşturarak bu sınıfa metodlar ve özellikler ekleyin ve bu class'ı kullanmaya çalışın.

*Bu sınıfı yazarken öğrendiğimiz özel metodların hepsini tanımlamaya çalışın.*

```python
class bilgisayar():
    
    def __init__(self, BS_durum ="Kapalı", BS_ses=0, BS_renk=0, Prog_List=["Windows"], Prog="Windows"):
        self.BS_durum = BS_durum
        self.BS_ses = BS_ses
        self.BS_renk = BS_renk
        self.Prog_List = Prog_List
        self.Prog = Prog
    
    def BS_ac(self):
        if(self.BS_durum=="Açık"):
            print("BS zaten açık durumda.")
        else :
            print("BS açılıyor...")
            self.BS_durum="Açık"
    
    def BS_kapat(self):
        if(self.BS_durum=="Kapalı"):
            print("BS zaten kapalı durumda.")
        else :
            print("BS kapanıyor...")
            self.BS_durum="Kapalı"
        
  
    def BS_ses(self):
        
        while True:
            cevap = input("Volume down: '<'\nVolume up: '>'\nExist : E")
            
            if(cevap=="<"):
                if (self.BS_ses !=0):
                    
                    self.BS_ses -=1
                    print("volume:", self.BS_ses)
                    
            elif (cevap==">"):
                if (self.BS_ses!=100):
                    
                    self.BS_ses +=1
                    print("volume", self.BS_ses)
            else:
                print("Volume updated:",self.BS_ses)
                break
                
    def BS_renk(self):
        
        while True:
            cevap = input("Color down: '<'\nColor up: '>'\nExist : E")
            
            if(cevap=="<"):
                if (self.BS_renk !=0):
                    
                    self.BS_renk -=1
                    print("color:", self.BS_renk)
                    
            elif (cevap==">"):
                if (self.BS_renk!=100):
                    
                    self.BS_renk +=1
                    print("color", self.BS_renk)
            else:
                print("color changed:",self.BS_renk)
                break
                
    def Prog_install(self, prog_name):
        print("Program installing...")
        time.sleep(1)
        
        self.Prog_name.append(prog_name)
        print("Program installed.")
    
    def Prog_select(self):   
         
        usedprog = input("İstediğiniz programı giriniz:")
        
        self.Prog = self.Prog_List[usedprog]
        print("Şu an kullanılan program:", self.Prog)
      
    
    def __len__(self):
        
        return len(self.Prog_List)
    
    def __str__(self):
    
        return "BS Durumu: {}\nBS Ses: {}\nProgram Listesi: {}\nUsed Prog: {}\n".format(self.BS_durum, self.BS_ses, self.Prog_List, self.Prog)
                    

BS = bilgisayar()

"""
Televizyon Uygulaması


1. BS Aç

2. BS Kapat

3. Ses Ayarları

4. Renk Ayarları

5. Prog Ekle

6. Prog Sayısını Öğrenme

7. Program Geçme

8. BS Bilgileri

Çıkmak için 'q' ya basın.
"""  
while True:

    işlem = input("İşlemi Seçiniz:")

    if (işlem == "q"):
        print("Program Sonlandırılıyor...")
        break

    elif (işlem == "1"):
        BS.BS_ac()
                    
    elif (işlem == "2"):
        BS.BS_kapat()

    elif (işlem == "3"):
        BS.BS_ses()

    elif (işlem == "4"):
        BS.BS_renk()
        
    elif (işlem == "5"):
        program_isimleri = input("Program isimlerini ',' ile ayırarak girin:")

        program_listesi = program_isimleri.split(",")

        for eklenecekler in program_listesi:
            BS.Program_install(eklenecekler)
   
    elif (işlem == "6"):

        print("Program Sayısı:", len(BS))

    elif (işlem == "7"):
        BS.Prog_select()
   
    elif (işlem == "8"):
        print(BS)

    else:
        print("Geçersiz İşlem......")       
        
```

### Proje 3 

Bu projede ise 4 tane sınıfı oluşturun. 

*Hayvan Sınıfı ------> Bütün hayvanların ortak özelliklerinin toplandığı sınıf olacak.*

*Köpek Sınıfı  ------> Bu sınıf, hayvan sınıfından miras alan bir sınıf olacak. Ayrıca bu sınıfa köpeklere ait ek özellikler ve metodlar ekleyin.*

*Kuş Sınıfı  ------> Bu sınıf, hayvan sınıfından miras alan bir sınıf olacak. Ayrıca bu sınıfa kuşlara ait ek özellikler ve metodlar ekleyin.*

*At Sınıfı   ------> Bu sınıf, hayvan sınıfından miras alan bir sınıf olacak. Ayrıca bu sınıfa atlara ait ek özellikler ve metodlar ekleyin.*
