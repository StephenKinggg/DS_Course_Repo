```python
# filter data of columns "sex","pclass","age","fare","survived"
titanic.loc[:,["sex","pclass","age","fare","survived"]]
```

```python
# draw graph below
new_tit= titanic.loc[:,["sex","pclass","age","fare","survived"]]
```

```python
# draw graph below
new_tit= titanic.loc[:,["sex","pclass","age","fare","survived"]]

g = sns.PairGrid(new_tit)

g = g.map_upper(sns.scatterplot, linewidths=1, edgecolor="w", s=40)
g = g.map_diag(sns.histplot)
g = g.map_lower(sns.scatterplot, linewidths=1, edgecolor="w", s=40)

g = g.add_legend();
```

```python
## 2.yöntem
g = sns.PairGrid(titanic, vars=["pclass","age","fare","survived"])

g = g.map_upper(sns.scatterplot, linewidths=1, edgecolor="w", s=40)
g = g.map_diag(sns.histplot)
g = g.map_lower(sns.scatterplot, linewidths=1, edgecolor="w", s=40)

g = g.add_legend();
```

```python
# Change default palette to GREEN and RED

g = sns.PairGrid(data=titanic,vars=["pclass","age","fare","survived"], hue="sex", palette={"female":"red","male":"green"})

g = g.map_upper(sns.scatterplot, linewidths=1, edgecolor="w", s=40)
g = g.map_diag(sns.kdeplot, shade=True)
g = g.map_lower(sns.scatterplot, linewidths=1, edgecolor="w", s=40)

g = g.add_legend();
```

3.1. What is the view? Why we use views?

Sql view öğeleri veritabanının altında tutulur. Bir yahut birden fazla tabloyu join ile birleştirip, where ile koşul belirttikten sonra içerisinde group by deyiminin kullanılabildiği sanal tabloya view denir. view in amaçlarından bir tanesi ise birden fazla tablodan çekilen verilerin tek tablo içerisinde toplanması ve bu sanal tablo içerisinden gerekli filtreler verilerek sorgu ile raporlama yapılabilmesidir, bunlar sql view avantajları arasında sayılabilir.

Sql view öğelerine ulaşabilmek için; veritabanı altında views dosyası genişletilir ve daha önce oluşturulmuş view ler görüntülenir.

Sql View Anahtar Kelimeleri ve Anlamları :

Create View : Yeni view oluşturmak için kullanılır.

Alter View : Daha önceden oluşturulmuş bir view üzerinde değişiklik yapılmak için kullanılır.

Drop View : Daha önceden oluşturulmuş bir view i veritabanından silmek için kullanılır.

With Encryption : View i şifrelemek için kullanılır.
