```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from collections import Counter
import warnings
warnings.filterwarnings('ignore') 
```

```python
labels = ["UK", "USA", "GER", "JAP", "FRA"]
age_men = [40, 35, 45, 53, 28]
age_women = [35, 32, 48, 48, 31]
```

```python
age = pd.DataFrame({"labels": labels,
                   "age_men": age_men,
                   "age_women": age_women})
age
```

***
# Two method: 
* Functional Method  
* Object Oriented Method
***

```python
plt.plot(labels, age_men)

plt.xlabel("labels")
plt.ylabel("age_men")

plt.title("Age means by Gender")

plt.show()
```

```python
# OOP

fig, ax = plt.subplots(2,1, figsize=(6,10))

ax[0].plot(labels, age_men, "r")
ax[0].set_xlabel("labels")
ax[0].set_ylabel("age_men")
ax[0].set_title("Age means by Gender")

ax[1].plot(labels, age_women, "b")
ax[1].set_xlabel("labels")
ax[1].set_ylabel("age_women")
ax[1].set_title("Age means by Gender")

plt.tight_layout()

plt.show()
```

***
Scatter plot , iki sayısal değişkenin karşılaştırılmasında kullandığımız bir yöntemdir ve direkt olarak değişkenler arasındaki korelasyon hakkında öngörü elde etmemizi sağlar. 

Barplot; nominal ve ordinal yani kategorik değişkenler içeren veri setlerinde verilerimizi ve gözlemleri kategorik gruplara ayırarak inceleme yapmamıza olanak sağlar. Pie chart, her gruba düşen yüzdeliği gösterirken, barplot ise her gruba düşen gözlemlerin sayısal değerlerini kullanır.

Piecharts; nominal ve ordinal yani kategorik değişkenler içeren veri setlerinde verilerimizi de gözlemleri kategorik gruplara ayırarak inceleme yapmamıza olanak sağlar.

Histogram; intervel ve ratio yani numerik değişkenler içeren veri setlerinin analizinde kullanılır.

Boxplot; intervel ve ratio yani numerik değişkenler içeren veri setlerinin analizinde ve Kategorilere göre bir değişkenin dağılımına bakarak whiskerların dışına düşen gözlemleri tespit etmek için kullanılır.

```python
labels = ["UK", "USA", "GER", "JAP", "FRA"]
age_men = [40, 35, 45, 53, 28]
age_women = [35, 32, 48, 48, 31]
```

```python
fig, ax = plt.subplots(figsize = (8,6))  # figsize ile ölçülerini genişlettik.

ax.bar(labels, age_men, width=0.25, label="men")
ax.set_title("Age means by Gender")
ax.set_ylabel("Age")


for p in ax.patches:
    ax.annotate(f"{p.get_height():.0f}", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', xytext =(0, 5), textcoords='offset points')

#for i in range(len(labels)):
#    plt.text(x=i, y=value[i], s=value[i], size=10, ha='center', va='center')

plt.legend()

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,6))  # figsize ile ölçülerini genişlettik.

ax.bar(labels, age_women, width=0.25, label = "women", color="red")
ax.set_title("Age means by Gender")
ax.set_ylabel("Age")


for p in ax.patches:
    ax.annotate(f"{p.get_height():.0f}", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', xytext =(0, 5), textcoords='offset points')

#for i in range(len(labels)):
 #   plt.text(x=i, y=value[i], s=value[i], size=9, ha='center', va='center')

plt.legend()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))

p = np.arange(len(labels))  # labels uzunluğu kadar bir array oluşturdum.
width = 0.4  # bir barın genişliği olarak atadığımız değer.

ax.bar(p - width/2, age_men,   width, label="men")  # p-width/2 dersek merkezden bunu bu kadar sola kaydır.
ax.bar(p + width/2, age_women, width, label="women")      # p+width/2 dersek merkezden bunu bu kadar sola kaydır.
ax.set_title("Age means by Gender")

ax.set_xticks(p)
ax.set_xticklabels(labels)

for p in ax.patches:
    ax.annotate(f"{p.get_height():.0f}", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext =(0, 5), textcoords='offset points')
    
plt.legend()

plt.show()
```
